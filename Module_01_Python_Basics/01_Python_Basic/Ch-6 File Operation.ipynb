### File Operation - Read and Write
File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

```python
# Read a Whole File

with open("Dummy_Files/Example.txt",'r') as file:
    content = file.read()
    print(content)
```

```python
# Read a file line by line

with open('Dummy_Files/Example.txt','r') as file:
    for line in file:
        print(line)
```

```python
# How to remove space between them

with open('Dummy_Files/Example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
# Writing a file(Overwriting)

with open('Dummy_Files/Example.txt','w') as file:
    file.write("Hello World!\n")
    file.write("This is a new line.\n")
    file.write("This is a second line")
    
```

```python
with open('Dummy_Files/Example.txt' , 'r') as file:
    print(file.read())
```

```python
# Write a file (without Overwriting)

with open('Dummy_Files/Example.txt','a') as file:
    file.write("\nAppend operation taking place!")
```

```python
# Writing a list of lines to a file

lines = ['\nFrist Line \n','Second Line \n','Third Line \n']
with open('Dummy_Files/Example.txt','a') as file:
    file.writelines(lines)
```

```python
# How to remove space between them

with open('Dummy_Files/Example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
# Writting Binary Files

data = b"\x00\x01\x02\x03\x04"

with open('Dummy_Files/B_Example.bin','wb') as file:
    file.write(data)
```

```python
# Reading Files 

with open('Dummy_Files/B_Example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
# Read the Content from source text file and write to a destination text file

# Copying a text files

with open("Dummy_Files/Example.txt",'r') as source_file:
    content = source_file.read()

with open("Dummy_Files/destination.txt",'a') as destination_file:
    destination_file.write(content)
```

## Working With File Paths
When working with files in Python, handling file paths correctly is crucial to ensure your code works across different operating systems and environments. Python provides several modules and functions for working with file paths effectively.

```python
#### Using the os module
import os
cwd=os.getcwd()
print(f"Current working directory is {cwd}")
```

```python
## create a new directory
new_directory="package"
os.mkdir(new_directory)
print(f"Directory '{new_directory}' create")
```

```python
## Listing Files And Directories
items=os.listdir('.')
print(items)
```

```python
### Joining Paths

dir_name="folder"
file_name="file.txt"
full_path=os.path.join(dir_name,file_name)
print(full_path)
```

```python
dir_name="folder"
file_name="file.txt"
full_path=os.path.join(os.getcwd(),dir_name,file_name)
print(full_path)
```

```python
path='mymodule.txt'
if os.path.exists(path):
    print(f"The path '{path}' exists")
else:
    print(f"The path '{path}' does not exists")
```

```python
#Checking if a Path is a File or Directory
import os

path = 'Ch-1. Python_Basic_print_function().ipynb'
if os.path.isfile(path):
    print(f"The path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory.")
else:
    print(f"The path '{path}' is neither a file nor a directory.")
```
