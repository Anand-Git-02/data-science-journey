# Tuples

A tuple in Python is similar to a list. The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.

Characterstics

- Ordered
- Unchangeble
- Allows duplicate

### Plan of attack

- Creating a Tuple
- Accessing items 
- Editing items
- Adding items
- Deleting items
- Operations on Tuples
- Tuple Functions

### Creating Tuples

```python
# empty
t1 = ()
print(t1)
type(t1)
```

```python
# create a tuple with a single item
t2 = ('hello',)
print(t2)
print(type(t2))
```

```python
# homo
t3 = (1,2,3,4)
print(t3)
# hetro
t4 = (1,2.5,True,[1,2,3])
print(t4)
```

```python
# tuple
t5 = (1,2,3,(4,5))
print(t5)
# using type conversion
t6 = tuple('hello')
print(t6)
```

### Accessing Items

- Indexing
- Slicing

```python
t3 = (1,2,3,4)

print(t3)
print(t3[0])
print(t3[-2])
```

```python
t5 = (1,2,3,(4,5))

t5[-1][1]
```

```python
t5 = (1,2,3,(4,5))

t5[0:2]

t5[::-1]
```

### Editing items

```python
t3 = (1,2,3,4)

print(t3)
t3[0] = 100
# immutable just like strings
```

### Adding items

```python
t3 = (1,2,3,4)
print(t3)
# not possible
```

### Deleting items

```python
t3 = (1,2,3,4)

print(t3)
del t3
print(t3)
```

```python
t = (1,2,3,4,5)
t[::-1]
```

```python
t = (1,2,3,4,5)

t[-1:-6:-1]
```

```python
print(t5)
del t5[-1]
```

### Operations on Tuples

```python
# + and *
t1 = (1,2,3,4)
t2 = (5,6,7,8)

print(t1 + t2)

print(t1*3)
```

```python
# membership
t1 = (1,2,3,4)

4 in t1
```

```python
t1 = (1,2,3,4,100)

# iteration
for i in t1:
  print(i)
```

### Tuple Functions

```python
# len/sum/min/max/sorted
t = (0,1,2,3,4,5,6)
len(t)
```

```python
t = (1,2,3)
sum(t)
```

```python
t1 = (1,2,3,4)
min(t)
```

```python
t1 = (1,2,3,4)
max(t1)
```

```python
t = (1,3,2,5,4)
sorted(t,reverse=False)
```

```python
# count

t = (1,2,3,4,5,5,5)

t.count(5)
```

```python
# index
t = (1,2,3,4,2,5,6,7,5,5)

t.index(2,2)
```

### Difference between Lists and Tuples

- Syntax
- Mutability
- Speed
- Memory
- Built in functionality
- Error prone
- Usability


```python
import time

L = list(range(10000000))
T = tuple(range(10000000))

start = time.time()

for i in L:
    i * 5
print(f"List time  :- {time.time()-start}")
 
start = time.time()
for i in T:
    i * 5
print(f"Tuple time :- {time.time()-start}")
```

```python
import sys

L = list()
T = tuple()

print(f'List  size :- {sys.getsizeof(L)}')

print(f'Tuple size :- {sys.getsizeof(T)}')
```

```python
base_size = 56
per_elements = 8
total_elements = 100

base_size + (per_elements * total_elements)
```

```python
import sys

L = list(range(100))
T = tuple(range(100))

print(f'List  size :- {sys.getsizeof(L)}')

print(f'Tuple size :- {sys.getsizeof(T)}')
```

```python
a = [1,2,3]
b = a

a.append(4)
print(a)
print(b)
```

```python
a = (1,2,3)
b = a

a = a + (4,)
print(a)
print(b)
```

### Special Syntax

```python
# tuple unpacking
a,b,c = (1,2,3)
print(a,b,c)
```

```python
a,b = (1,2,3)
print(a,b)
```

```python
a = 1
b = 2
a,b = b,a

print(a,b)
```

```python
a,b,*others = (1,2,3,4,7,8,89,)
print(a,b)
print(others)
```

```python
# zipping tuples
a = (1,2,3,4)
b = (5,6,7,8)

list(zip(a,b))
```
