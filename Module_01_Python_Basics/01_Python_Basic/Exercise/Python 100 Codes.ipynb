### 2. Write a program that will convert celsius value to fahrenheit


```python
# °F = (°C × 9/5) + 32
# °C = (°F - 32) x 5/9

def fahrenheit(num):

    F  = (num * (9/5)) + 32

    return F


def celsius(num):

    C = (num - 32) * (5/9)

    return C


print("Fahrenheit :-",fahrenheit(100))
print("Celsius    :-",celsius(212))
```

### 4. Write a program that will give you the sum of 3 digits

```python
def Sum(*numbers):
    Result = 0
    for i in numbers:
        Result += i

    return Result

Sum(1,2,3)
```

### 5. Write a program that will reverse a four digit number.Also it checks whether the reverse is true.

```python
def reverse_num(number):
    Result = ''
    for i in str(number):
        Result = i+Result

    if str(number)[::-1] == Result:
        print("The reverse is true")
        return int(Result)
        
reverse_num(1234)
```

### 6. Write a program that will tell whether the number entered by the user is odd or even.

```python
def is_even():

    try :
        num = int(input("Enter Your Number :-"))
    
        if num % 2 == 0:
            return f"{num} is Even"
        else:
            return f"{num} is Odd"
    except ValueError:
        return "Invalid Input"
        


print(is_even())
```

### 7. Write a program that will tell whether the given year is a leap year or not.

```python
def is_leap_year():
    year = int(input("Enter Your Year:-"))

    if year % 4 == 0:
        if year % 100 == 0:
            return f"{year} is not leap year"
        else:
            return f"{year} is leap year"
    else:
        return f"{year} is not leap year"

print(is_leap_year())
```

### 8. Write a program to find the euclidean distance between two coordinates.

```python
def Euclidean_distance():

    try:
        x1 , x2 = map(float,input("Enter Your X1 , X2 Coordinates").split(","))
        y1 , y2 = map(float,input("Enter Your Y1 , Y2 Coordinates").split(","))

        diff_x = (x2 - x1)
        diff_y = (y2 - y1)

        d = ((diff_x**2) + (diff_y**2)) ** 0.5

        return f"Distance Between two coordinates : {d:.3f}"
    
    except (ValueError, TypeError):
        return "Invalid Input"

print(Euclidean_distance())
```

### 9. Write a program that take a user inputr of three angles and will find out whether it can form a triangle or not.

```python
def is_triangle():
    try:
        a,b,c = map(float,input("Enter Your Angles a, b, c (comma-separated): ").split(','))

        total = a + b + c

        if total == 180:
            return f"It can form a Triangle"

        else:
            return f"It can't form a Triangle"
            
    except (ValueError, TypeError):
        return "Invalid Input"
        
        
print(is_triangle())
```

### 10. Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit

```python
def is_profit():
    try:
        cost_price = float(input("Enter Your Cost Price :-"))
        sell_price = float(input("Enter Your Sell Price :-"))

        if cost_price > sell_price:
            return f"Your Loss is {cost_price - sell_price}"

        elif sell_price > cost_price:
            return  f"Your Profit is {sell_price - cost_price}"

        else:
            return  "No Profit,No Loss"
            
    except (ValueError, TypeError):
        return "Invalid Input"

print(is_profit())
```

### 11. Write a program to find the simple interest when the value of principle,rate of interest and time period is given. 

```python
def simple_interest():
    try:
        P = float(input("Enter the Principal Amount             : "))
        R = float(input("Enter the Rate of Interest (% per year): "))
        T = float(input("Enter the Time Period (years)          : "))

        if P < 0 or R < 0 or T < 0:
            return "Values cannot be negative."
            

        SI = (P*R*T)/100
        return f"Your Simple Interest is: {SI:.2f}"
        
    except (ValueError, TypeError):
        return "Invalid input. Please enter valid numbers."

print(simple_interest())
```
