1. Write a program to swap two variables without using a third variable.
   **Input:** `a = 5, b = 10`
   **Output:** `a = 10, b = 5`

```python
a = 5
b = 10

a , b = b , a

print(f"a = {a}, b = {b}")
```

9. Convert seconds into hours, minutes, and seconds.
   **Input:** `seconds = 3660`
   **Output:** `1 hour, 1 minute, and 0 seconds`



```python
# Input: total seconds
seconds = 3660

# Convert seconds to hours, minutes, and remaining seconds
hours = seconds // 3600
minutes = (seconds % 3600) // 60
remaining_seconds = seconds % 60

# Handle singular/plural forms
hour_str = "hour" if hours == 1 else "hours"
minute_str = "minute" if minutes == 1 else "minutes"
second_str = "second" if remaining_seconds == 1 else "seconds"

# Output the result
print(f"{hours} {hour_str}, {minutes} {minute_str}, and {remaining_seconds} {second_str}")
```

12. Calculate the sum of digits of a three-digit number.
    **Input:** `number = 345`
    **Output:** `Sum of digits = 12`



```python
number = input("Enter Three Digit Number")

Sum_num = 0

for i in range(len(number)):
    Sum_num = Sum_num + int(number[i])
    
print(f"Sum of digits = {Sum_num}")
```
