### Exercise 1: Nested Data Structures
**Task:** Convert all strings in a nested list of tuples and strings to uppercase and flatten the structure into a single list.

**Steps:**
1. Iterate through each element of the outer list.
2. Check if the element is a tuple or list.
3. If it's a tuple or list, iterate through its elements.
4. Convert strings to uppercase.
5. Append elements to a new list.
6. Flatten the structure.

**Expected Output:**  
Input: `[("apple", "banana"), ["carrot", ("date", "elderberry")]]`  
Output: `['APPLE', 'BANANA', 'CARROT', 'DATE', 'ELDERBERRY']`
