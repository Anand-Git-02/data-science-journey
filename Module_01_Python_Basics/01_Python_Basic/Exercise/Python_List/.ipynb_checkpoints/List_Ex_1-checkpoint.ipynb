
### Easy Level

1. **Sum of Elements in a List**
   - **Input:** A list of integers.
   - **Task:** Find the sum of all elements in the list.
   - **Input Example:** `[2, 4, 6, 8]`
   - **Output Example:** `20`

```python
Input = [2,4,6,8]
sum(Input)
```

```python
Input = [2,4,6,8]

Sum = 0

for i in Input:
    Sum = Sum + i
print(Sum)
```

```python
Input = [2,4,6,8]

Sum = 0

for i in range(len(Input)):
    Sum = Sum + Input[i]
print(Sum)
```

   - **Input:** A list of elements.

   - **Input Example:** `[3, 2, 1, 4]`

```python
Input = [3, 2, 1, 4]

Input[::-1]
```

```python
Input = [3, 2, 1, 4]
Outpt = []
for i in range(len(Input)-1,-1,-1):
    Outpt.append(Input[i])
Outpt
```

10. **Concatenate Two Lists**
   - **Input:** Two lists.
   - **Task:** Concatenate them into a single list.
   - **Input Example:** `[1, 2]`, `[3, 4]`
   - **Output Example:** `[1, 2, 3, 4]`

```python
a = [1, 2] 
b = [3, 4]

Output = []

for i in range(2):
    if a[i] not in Output:
        Output.append(a[i])
for j in range(2):
    if b[j] not in Output:
        Output.append(b[j])
        
Output
```
