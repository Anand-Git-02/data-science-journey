### 1. **Sum of Digits in an Integer**
   - **Task**: Write a function `sum_of_digits(n)` that takes an integer and returns the sum of its digits.
   - **Input**: `1234`
   - **Output**: `10`

```python
def sum_of_digits(n):
    Sum = 0
    for i in n:
        Sum = Sum + int(i)
    return Sum    

n = input("Enter Your Number :-")
sum_of_digits(n)
```

### 2. **Find Common Elements in Two Lists**
   - **Task**: Write a function `common_elements(list1, list2)` that returns a list of common elements between two lists.
   - **Input**: `[1, 2, 3, 4], [3, 4, 5, 6]`
   - **Output**: `[3, 4]`



```python
def common_elements(list1, list2):
    Output = []
    for i in list1:
        for j in list2:
            if i == j:
                if i not in Output:
                    Output.append(i)
    return Output


L1 = [1, 2, 3, 4,8]
L2 = [3, 4, 3, 6,8]

common_elements(L1,L2)
```

### 5. **Find the GCD of Two Numbers**
   - **Task**: Write a function `gcd(a, b)` that returns the greatest common divisor (GCD) of two numbers.
   - **Input**: `8, 12`
   - **Output**: `4`




```python
def gcd(a,b):
    Divisor_a = 0
    Divisor_b = 0
    
    for i in range(1,a+1):
        if (a % i == 0):
            Divisor_a = Divisor_a + 1
    for j in range(1,b+1):
        if (a % i == 0):
            Divisor_b = Divisor_b + 1
    return Divisor_a,Divisor_b
        
gcd(8,12)
```

### 6. **Find the LCM of Two Numbers**
   - **Task**: Write a function `lcm(a, b)` that returns the least common multiple (LCM) of two numbers.
   - **Input**: `4, 5`
   - **Output**: `20`


To compute the least common multiple (LCM) of two numbers, you can use the relationship between LCM and the greatest common divisor (GCD). The formula is:

Lcm(a,b) = |axb|/GCD(a,b)


```python
def gcd(a,b):

    while b != 0:
        a,b  = b, a % b
    return a
    
def Lcm(a,b):
    LCM = abs(a*b)//gcd(a,b)
    return LCM
Lcm(4,5)
```
