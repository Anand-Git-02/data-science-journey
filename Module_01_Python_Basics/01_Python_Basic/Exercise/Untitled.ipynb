```python
vowel = "aeiou"

string = "Education"

vowel_count = 0

for i in string.lower():

    if i in vowel:
        vowel_count += 1
print(vowel_count)
```

```python
import pandas as pd
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
num_samples = 10  # Number of rows in the dataset

# Generate random numbers for num1 and num2
data = {
    'num1': np.random.randint(1, 100, num_samples),
    'num2': np.random.randint(1, 100, num_samples)
}

# Calculate the sum of num1 and num2
data['Target'] = data['num1'] + data['num2']

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```
