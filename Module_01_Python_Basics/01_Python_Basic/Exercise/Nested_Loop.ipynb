### What is a Nested For Loop?

A **nested for loop** is simply a loop inside another loop. The inner loop will run entirely each time the outer loop runs one iteration. This is useful when you need to iterate over multi-dimensional data, such as lists of lists (like matrices or tables).

```python
# Outer loop
for i in range(3):
    # Inner loop
    for j in range(2):
        print(f"i = {i}, j = {j}")
```

### Explanation:
1. The outer loop (`for i in range(3)`) will iterate 3 times (with `i` taking values 0, 1, 2).
2. The inner loop (`for j in range(2)`) will run 2 times for each iteration of the outer loop (with `j` taking values 0, 1).

```python
### Practical Example: Working with 2D Lists (Matrices)

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Outer loop to iterate over rows
for row in matrix:
    # Inner loop to iterate over columns (elements in each row)
    for element in row:
        print(element, end=" ")
    print()  # Print a newline after each row
```

### Explanation:
- The outer loop iterates through each row of the matrix.
- The inner loop goes through each element in the current row.
- We print each element in the same row on the same line (`end=" "` ensures this), and `print()` is used to move to the next line after each row.

```python
### Nested Loop Pattern Example

n = 5
for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    print()  # Move to the next line after each row
```
