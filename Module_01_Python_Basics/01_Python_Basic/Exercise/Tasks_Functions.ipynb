### 1. **Phonebook Management System**
Write a phonebook management system with the following functionality:
- Add a new contact with a name and phone number.
- Search for a contact by name or phone number.
- Delete a contact.
- Display all contacts.
  
You must implement this as a function-based solution using lists and tuples to store contacts.

**Input:**  
```
add_contact("Alice", "12345")
add_contact("Bob", "67890")
search_contact("Alice")
delete_contact("Bob")
display_contacts()
```

**Output:**  
```
Contacts:
Alice: 12345
```

### 2. **Library Book Management**
Write a function that manages books in a library. The function should:
- Add books to the library (each book represented as a dictionary with title, author, and year).
- Remove a book by title.
- Search for books by author.
- Display all books sorted by year.

**Input:**  
```
add_book("1984", "George Orwell", 1949)
add_book("Brave New World", "Aldous Huxley", 1932)
remove_book("1984")
search_by_author("Aldous Huxley")
display_books_sorted_by_year()
```

**Output:**  
```
Books:
Title: Brave New World, Author: Aldous Huxley, Year: 1932
```

### 3. **Inventory Management System**

### 4. **Student Grade Management**
Write a function to manage student grades. The system should:
- Allow adding students with their grades in different subjects.
- Calculate the average grade for each student.
- Identify the student with the highest average.
- List students who are failing (average grade below 40).

**Input:**  
```
add_student("Alice", {"Math": 85, "Science": 90, "English": 80})
add_student("Bob", {"Math": 30, "Science": 35, "English": 40})
add_student("Charlie", {"Math": 95, "Science": 100, "English": 90})
average_grade("Alice")
find_top_student()
list_failing_students()
```

**Output:**  
```
Average grade for Alice: 85
Top student: Charlie
Failing students: Bob
```
