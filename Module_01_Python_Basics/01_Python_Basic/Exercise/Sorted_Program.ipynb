```python
a = [5,6,3,2,1,4]

for i in range(0 , len(a)-1):
    min_index = i
    for j in range(i+1, len(a)):
        if a[j] < a[min_index]:
            min_index = j
    if min_index != i:
        a[i] , a[min_index] = a[min_index] , a[i]
a
```

### Bubble Sort

```python
a = [5,6,3,2,1,4]

# Outer Loop : i = 0 to n-1
# inner Loop : j = 0 to n-1-i

for i in range(0,len(a) - 1):
    for j in range(0,len(a)-1-i):
        if a[j] > a[j+1]:
            a[j] , a[j+1] = a[j+1] , a[j]
a
```
