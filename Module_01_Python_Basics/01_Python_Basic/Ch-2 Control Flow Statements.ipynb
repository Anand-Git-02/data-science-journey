## 2.1 Decision-Making Structures
- Decision-making structure require the progrmmer to specify one or more coditions to be evaluted or tested by the program,along with a statement or statements to be executed if the codition is determined to be true, and optionally, other statement to be executed if the codition is determine to be false.
- The "if" structure is the genneral from of a typical decision-making structure found in most of the programming languages.
- When we need to execute a set of statements based on a condition , then we need to use the "if" structure.
- There are four types of "if" statements that we can use in Python programming based on the requirement:
- "If" Statement,
- If...else Statement,
- Nested "if" Statement, and
- If-elif-else Ladder.

### 2.2.1 For loop 

- If the number of repetition are known in advance then a "for" loop can be used.
- For executing for loop we need to specify
- **start**
- **end**
- **step argument**

#### syntax

for var in range(start, end, step):
   - command to be exected
##### where
- **var** is the variable name
- **start** denotes the starting number
- **end** ddenoted the ending number
- **step** is the optional step for increment or decrement (**default are +1 and -1**)


```python
# program for printing the table of number.


num = int(input("Enter the number to print the table: "))

for k in range(1,11):
    print(num,"x", k,"=",num*k)

# Outside the for loop

print("Finished Task")
```

```python
# Program to show use of "For" and "If" in the same program.

num = int(input("Enter the number of Item : "))

amount = 0
discount = 0

# Using for loop

for k in range(1,num+1):
    
    print(f"Item {k}")
    
    price = eval(input("Enter the price : "))
    
    quantity = eval(input("Enter the quantity : "))
    
    amount = amount + (price*quantity)
    
    
print("Total is : ",amount)

# Using "if" coditional statement.

if amount > 5000:
    
    discount = amount * 0.1
    print("Discount of 10% = ", discount)

elif amount > 3000:
    
    discount = amount * 0.07
    print("Discount of 7% = ", discount)

elif amount > 1000:
    
    discount = amount * 0.05
    print("Discount of 5% = ", discount)
    
else:
    print("No discount received")
    
# In main program

print("Net amount is : ", amount-discount)
print("Finished Task")
```

### 2.3 Nesting of Conditional Statements and Loop 
It is also possible to nest conditional statements and different loops. For example, conditional statement can be inside for/while loop or for/while loop can be inside conditional statements.

```python
# Program for printing even/odd numbers depending on user's choice.

choice = int(input("Enter Your Choice: 1 for odd, 2 for even: "))

# Using for inside if structure

if choice == 1:
    last = int(input("Enter Your Last Number : "))
    
    for k in range(1,last+1,2):
        print(k)
        

elif choice == 2:
    last = int(input("Enter Your Last Number : "))
    
    for k in range(2,last+1,2):
        print(k)
else:
    print("Invalid Choice")
    
```

### 2.3.2 The 'if' Conditional Staement Inside 'for' Loop

```python
# Program to compute the length of given string ignoring spaces.

string = input("Enter the String: ")
count = 0
space_count = 0

# Use of if inside for loop

for letter in string.strip():
    if letter != " ":
        count +=  1
    else:
        space_count += 1
        
# Outside the loop.
print("The length of the string is:",count)
print("The length of the Space is:",space_count)
```

```python
# Program to print all different arrangements of the letter M, N and O.

for first in "MNO":
    for second in "MNO":
        if second != first:
            for third in "MNO":
                if third != first and third != second:
                    print(first + second + third)
```

```python
import random

Jackpot = random.randint(1,100)

attempts = 1

Guess_number = int(input("Guess the Number : "))


while Guess_number != Jackpot:

    if Guess_number > Jackpot:
        print("wrong! Guess Lower\n")
    
    else:
        print('wrong! Guess Higher\n')
        
    Guess_number = int(input("Guess the Number : "))
    
    attempts = attempts + 1
    
else:
    print("Correct Guess\n")
    print('Attempts',attempts)
```

```python
import random

Jackpot = random.randint(1,100)

Guess_num = int(input("Enter your Number :-"))

attempts = 1

while Jackpot != Guess_num:
    
    if Guess_num > Jackpot:
        print("Guess Lower!!")
    elif Guess_num < Jackpot:
        print("Guess Higher!!")
    else:
        print("Enter valid Number")
    Guess_num = int(input("Enter Your Number :"))
    attempts = attempts + 1

else:
    print("You Win!!")
```
