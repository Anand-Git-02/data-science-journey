```python
def is_even(num):
    """
    This function returns if a given number is odd or even
    input - any valid integer
    output - odd/even
    created on - 21 Aug 2024
    """
    if type(num) == int:
        if num % 2 == 0:
            return f"{num} is Even"
    
        else:
            return f"{num} is Odd"
    else:
        return "Invaild Input"

is_even(25)
```
