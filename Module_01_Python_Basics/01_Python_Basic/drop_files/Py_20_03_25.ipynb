1. **Task: Count Positive and Negative Numbers**  
   - **Input:** `[-2, -1, 0, 1, 2, 3]`  
   - **Output:** `Positive: 3, Negative: 2`

```python
User_List = [-2, -1, 0, 1, 2, 3,4]

Positive = 0
Negative = 0
Zero = 0

for i in User_List:
    if i > 0:
        Positive += 1
    elif i < 0:
        Negative += 1
    else:
        Zero += 1

print(f"Positive: {Positive}")
print(f"Negative: {Negative}")
print(f"Zero    : {Zero}")
```
