### 4. **Generate Prime Numbers up to n**
   Write a function that takes an integer `n` and returns a list of all prime numbers up to `n`.
   - **Input**: `n = 20`
   - **Output**: `[2, 3, 5, 7, 11, 13, 17, 19]`



```python
6n + 1   
    
def primes_up_to_n(n):
    def is_prime(num):
        
        for i in range(3, (num // 2) + 1, 2):
            if num % i == 0:
                return False
        return True
    
    prime_list = []
    for i in range(2, n + 1):
        if is_prime(i):
            prime_list.append(i)
    
    return prime_list

# Example usage
n = 20
print(primes_up_to_n(n))  # Output: [2, 3, 5, 7, 11, 13, 17, 19]
```

```python
def primes_up_to_n(n):
    primes = []
    
    for num in range(2, n + 1):
        
        is_prime = True
        
        for i in range(2, num):
            
            if num % i == 0:
                is_prime = False
                break
                
        if is_prime:
            primes.append(num)
    
    return primes

# Example usage
n = 20
print(primes_up_to_n(n))  # Output: [2, 3, 5, 7, 11, 13, 17, 19]
```

### 6. **Flatten a Nested List**
   Write a function that takes a nested list and returns a flattened version of the list.
   - **Input**: `[[1, 2], [3, 4], [5, [6, 7]]]`
   - **Output**: `[1, 2, 3, 4, 5, 6, 7]`



```python
a = [[1, 2], [3, 4], [5, [6, 7]]]
b = []

for i in a:
    for j in i:
        if type(j) == int:
            b.append(j)
        elif type(j) == list:
            for ele in j:
                b.append(ele)
b
```

### 10. **Check if a Number is a Perfect Square**
   Write a function to check if a given number is a perfect square. A perfect square is an integer that is the square of another integer.
   - **Input**: `n = 16`
   - **Output**: `True` (since \(4^2 = 16\))

```python
N = int(input())

my_list = []

for i in range(N):
    
    command = input("Enter Your Operation").strip().split()
    
    if command[0] == 'append':
        value = int(command[1])
        my_list.append(command[1])
    
    elif command[0] == 'insert':
        index = int(command[1])
        value = int(command[2])
        my_list.insert(index,value)
        
    elif command[0] == 'pop':
        my_list.pop()
    
    elif command[0] == 'reverse':
        my_list.reverse()
        
    elif command[0] == 'sort':
        my_list.sort()
        
    elif command[0] == 'remove':
        value = int(command[1])
        my_list.remove(value)
        
    elif command[0] == 'print':
        print(my_list)
```
