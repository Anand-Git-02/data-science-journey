```python
string = "1 w 2 d 3e"

list_char = string.split()

output = []

for char in list_char:
    print(char)
    
    if char.isalpha():
        output.append(char.capitalize())
    else:
        output.append(char)
        
result = " ".join(output)

print(result)
```

```python
def capitalize_name(full_name):
    # Split the input string into words
    words = full_name.split()
    
    # Initialize an empty list to hold capitalized words
    capitalized_words = []
    
    # Loop through each word in the list
    for word in words:
        # Check if the first character is alphabetic
        if word[0].isalpha():
            # Capitalize the first letter and keep the rest as is
            capitalized_word = word[0].upper() + word[1:]
        else:
            # Leave the word unchanged if the first character is not alphabetic
            capitalized_word = word
        
        # Add the capitalized word to the list
        capitalized_words.append(capitalized_word)
    
    # Join the capitalized words back into a single string
    capitalized_name = ' '.join(capitalized_words)
    return capitalized_name

# Input
full_name = input().strip()

# Output
print(capitalize_name(full_name))
```
