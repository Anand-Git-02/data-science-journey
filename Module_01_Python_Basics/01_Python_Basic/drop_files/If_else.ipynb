### Exercise 3: Grade Classification
1. Ask the user to enter a grade (0-100).
2. Check if the grade is 90 or higher.
3. If it is, print that the grade is A.
4. If it isn't, check if the grade is 80 or higher.
5. If it is, print that the grade is B.
6. If it isn't, check if the grade is 70 or higher.
7. If it is, print that the grade is C.
8. If it isn't, check if the grade is 60 or higher.
9. If it is, print that the grade is D.
10. If it isn't, print that the grade is F.

```python
Grade = int(input("Enter a grade (0-100) :-"))

if Grade >= 90 or Grade > 90:
    print("Grade is A")
elif Grade == 80 or Grade > 80:
    print("Grade is B")
elif Grade >= 70 or Grade > 70:
    print("Grade is C")
elif Grade == 60 or Grade > 60:
    print("Grade is D")
else:
    print("Grade is F")
```

### Exercise 4: Leap Year Checker
1. Ask the user to enter a year.
2. Check if the year is divisible by 4.
3. If it is, check if the year is not divisible by 100.
4. If it is not divisible by 100, print that the year is a leap year.
5. If it is divisible by 100, check if it is also divisible by 400.
6. If it is divisible by 400, print that the year is a leap year.
7. If it isn't divisible by 400, print that the year is not a leap year.
8. If the year is not divisible by 4, print that the year is not a leap year.

```python
# 2020
# 2024
# 2028
# 2032
# 2036
# 2040
# 2044
# 2048
# 2052
# 2056
```

```python
year = int(input("Enter Your Year:-"))

if year % 4 == 0:
    if year % 100 != 0:
        print(f"{year} is a leap Year")
    else:
        if year % 400 == 0:
            print(f"{year} is not a leap Year")
else:
    print(f"{year} is not a leap Year")
```

### Exercise 5: Vowel or Consonant
1. Ask the user to enter a letter.
2. Convert the letter to lowercase to handle both uppercase and lowercase input.
3. Check if the letter is one of 'a', 'e', 'i', 'o', 'u'.
4. If it is, print that the letter is a vowel.
5. If it isn't, print that the letter is a consonant.

```python
word = input("Enter Your Word :-").lower()

if "a" in word:
    print("letter is a vowel")

elif "e" in word:
    print("letter is a vowel")

elif "i" in word:
    print("letter is a vowel")

elif "o" in word:
    print("letter is a vowel")

elif "u" in word:
    print("letter is a vowel")

else:
    print("letter is a consonant.")
```

```python
image = [[1,1,0],
           [1,0,1],
           [0,0,0]]

for i in range(len(image)):
    image[i] = image[i][::-1]
    
    for j in range(len(image[i])):
        if image[i][j] == 0:
            image[i][j] = 1
        else:
            image[i][j] = 0
print(image)
```
