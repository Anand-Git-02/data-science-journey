1. **Task: Count Positive and Negative Numbers**  
   - **Input:** `[-2, -1, 0, 1, 2, 3]`  
   - **Output:** `Positive: 3, Negative: 2`

```python
List_1 = [-2, -1, 0, 1, 2, 3]

Negative = 0
Postive = 0

for i in List_1:
    if i < 0:
        Negative = Negative + 1
    elif i > 0:
        Postive = Postive + 1
print(f"Postive  :- {Postive}")
print(f"Negative :- {Negative}")
```

3. **Task: Find the Maximum in a List**  
   - **Input:** `[5, 2, 9, 1, 7]`  
   - **Output:** `Maximum: 9`

```python
List_2 = [5, 2, 9, 1, 7,100]

Maximum = List_2[0]

for i in List_2:
    if i > Maximum:
        Maximum = i
Maximum
```

5. **Task: Count Words in a String**  
   - **Input:** `"hello world this is python"`  
   - **Output:** `Number of words: 5`

```python
string_1 = "hello world this is python"

word_count = 1

for i in string_1.strip():
    if i == " ":
        word_count += 1
word_count
```

8. **Remove Duplicates**
   - **Input:** A list of integers.
   - **Task:** Remove any duplicate values from the list.
   - **Input Example:** `[1, 2, 2, 3, 4, 4]`
   - **Output Example:** `[1, 2, 3, 4]`

```python
L3 = [1, 2, 2, 3, 4, 4]
list(set(L3))
```

```python
L4 = [1, 2, 2, 3, 4, 4]

no_dupicate = []

for i in L4:
    if i not in no_dupicate:
        no_dupicate.append(i)
```
