1. **Recursive Sum of Digits**  
   Write a function that takes a number and recursively returns the sum of its digits until it's a single digit.  
   **Example:**  
   ```python
   Input: 9875  
   Output: 2  (9+8+7+5 = 29 → 2+9 = 11 → 1+1 = 2)
   ```

````python
def recursive_sum(n):

    if n < 10:
        return n

    digit_sum = 0
    for i in str(n):
        digit_sum = digit_sum + int(i)

    return recursive_sum(digit_sum)

recursive_sum(9875)
````

2. **Spiral Matrix**  
   Given a `N x N` matrix, return all its elements in spiral order.  
   **Example:**  
   ```python
   Input: [[1, 2, 3],  
           [4, 5, 6],  
           [7, 8, 9]]  
   Output: [1, 2, 3, 6, 9, 8, 7, 4, 5]  
   ```



````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

result = []

top = 0
bottom = len(matrix) - 1

left = 0
right = len(matrix[0]) - 1

while top <= bottom and left <= right:
    # Move right
    for i in range(left , right + 1):
        result.append(matrix[top][i])
    top += 1
    
    
    
````

   ```python
   Input: [100, 4, 200, 1, 3, 2]
   ```

8. **Kth Largest Element**  
   Given a list, find the **Kth largest element**.  
   **Example:**  
   ```python
   Input: [3, 2, 1, 5, 6, 4], K=2  
   Output: 5  
   ```



````python
def Kth_largest(K,list_1):
    Sorted_list = sorted(list_1,reverse=True)
    
    return Sorted_list[K-1]

list_1 = [3, 2, 1, 5, 6, 4]
Kth_largest(2,list_1)
````

9. **Balanced Parentheses**  
   Check if a given string containing `()`, `{}`, `[]` is **balanced**.  
   **Example:**  
   ```python
   Input: "{[()()]}"  
   Output: True  
   ```

````python
string = "{[()()]}"

start_bkt = 0

end_bkt = 0

for i in string:
    if i in ["(","{","["]:
        start_bkt = start_bkt + 1

    elif i in [")","}","]"]:
        end_bkt = end_bkt + 1

if start_bkt == end_bkt:
    print(True)
else:
    print(False)
````
