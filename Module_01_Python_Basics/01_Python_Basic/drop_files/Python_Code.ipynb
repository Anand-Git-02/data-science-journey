```python
n = 15

for i in range(1,n):
    for j in range(1,n):
        if j>=n-i:
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
def pyramid_pattern(n):
    # n = number of rows
    for i in range(1, n+1):
        for j in range(1, 2*n):  # total columns = 2*n - 1
            if (j >= n+1-i) and (j <= n-1+i):
                print("*", end="")
            else:
                print(" ", end="")
        print()

# Example usage
pyramid_pattern(10)
```
