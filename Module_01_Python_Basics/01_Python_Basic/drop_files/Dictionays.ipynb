### 3. **Extract Unique Values from a List of Dictionaries**
   - **What to Do:** Given a list of dictionaries, extract all unique values for a specific key.
   - **Input Example:** ` [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 20}, {'name': 'Charlie', 'age': 25}]`, extract unique ages.
   - **Output Example:** `[25, 20]`



```python
Input = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 20}, {'name': 'Charlie', 'age': 25}]

Output = []

for i in Input:
    print(i['age'])
    Output.append(i['age'])
    
Output = set(Output)
list(Output)
```

### 4. **Group Dictionaries by a Key Value**
   - **What to Do:** Given a list of dictionaries, group them by a specific key.
   - **Input Example:** ` [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 20}, {'name': 'Alice', 'age': 30}]`, group by `name`.
   - **Output Example:** ` {'Alice': [{'name': 'Alice', 'age': 25}, {'name': 'Alice', 'age': 30}], 'Bob': [{'name': 'Bob', 'age': 20}]}`
   

```python
Input = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 20}, {'name': 'Alice', 'age': 30}]
list_items_1 = []
list_items_2 = []

for i in Input:
#     print(i)
    
    if i['name'] == 'Alice':
        list_items_1.append(i)

    elif i['name'] == 'Bob':
        list_items_2.append(i)
        
dic_1 = {Input[0]['name']:list_items_1}
dic_2 = {Input[1]['name']:list_items_2}

dic_1.update(dic_2)
        
print(dic_1)
```

### 6. **Calculate the Average of a Key's Values**
   - **What to Do:** Given a list of dictionaries, calculate the average of the values for a specific key.
   - **Input Example:** ` [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 20}, {'name': 'Charlie', 'age': 30}]`, calculate the average age.
   - **Output Example:** `25`


```python
Input = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 20}, {'name': 'Charlie', 'age': 30}]

Output = []

for i in Input:
#     print(i['age'])
    
    Output.append(i['age'])
    
avg_of_age = sum(Output)/3
avg_of_age
```
