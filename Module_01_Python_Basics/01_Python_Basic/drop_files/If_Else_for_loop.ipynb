### Calulator Program

```python
num1 = int(input("Enter Your Number 1 :- "))

operation = input('Operations(+-x/)    :- ')

num2 = int(input("Enter Your Number 2 :- "))

if operation == '+':
    Addition = num1 + num2
    print(f'{num1} {operation} {num2} = {Addition}')

elif operation == '-':
    subtration = num1 - num2
    print(f'{num1} {operation} {num2} = {subtration}')

elif operation == 'x':
    multiplication = num1 * num2
    print(f'{num1} {operation} {num2} = {multiplication}')

elif operation == '/':
    division = num1 / num2
    print(f'{num1} {operation} {num2} = {division}')

else:
    print("Error Invaild Opertor")
    
```

1. **Task: Count Positive and Negative Numbers**  
   - **Input:** `[-2, -1, 0, 1, 2, 3]`  
   - **Output:** `Positive: 3, Negative: 2`



```python
# Step :- 1 Create List
List1 = [-2, -1, 0, 1, 2, 3]

# Count of Positive and Negative
Positive = 0
Negative = 0

# Step :- 2 Iteraing Elements from the list 
# dividing elements on basis of condition 
for i in List1:
    # print(List1[i])
    if i > 0:
        Positive = Positive + 1
    elif i < 0:
        Negative = Negative + 1

print(f"Positive :- {Positive}")
print(f"Negative :- {Negative}")
```

9. **Task: Count Vowels in a String**  
   - **Input:** `"education"`  
   - **Output:** `Number of vowels: 5`

```python
str1 = "education"
vowels = "aeiou"

count = 0

for i in str1:
    if i in vowels:
        count += 1

print(count)
```

10. **Task: Check if a Number is Prime**  
    - **Input:** `7`  
    - **Output:** `7 is a prime number`

```python
n = 7

for i in range(1,n+1):
    if i % 2 ==0:
        print(f"{n} is not prime")
        break
else:
    print(f"{n} is prime")
```
