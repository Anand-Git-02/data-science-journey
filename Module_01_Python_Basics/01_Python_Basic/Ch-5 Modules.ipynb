### Introduction to Modules in Python

In Python, a **module** is a file containing Python code (functions, variables, and classes) that can be reused in other programs. Modules allow you to organize your code into separate files, making it easier to maintain, read, and reuse. Instead of writing everything in one script, you can split related functionality into different modules, which can then be imported into other Python files.

#### Why Use Modules?
1. **Code Reusability**: Write functions once and use them in multiple programs.
2. **Maintainability**: Organize code logically, making it easier to maintain.
3. **Namespace Management**: Avoid naming conflicts by encapsulating code into modules.
4. **Built-in Functions**: Python provides numerous built-in modules, such as `math`, `os`, `random`, and `datetime`, offering powerful pre-written functionality.

#### How to Create and Use a Module

A module is simply a `.py` file containing Python code. Here's an example:

##### Creating a Module:
Let's say you have a file named `math_operations.py` with the following content:

```python
# math_operations.py
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
```

##### Importing a Module:
To use the functions defined in the module, you can import it into another Python script:

```python
# main.py
import math_operations

result = math_operations.add(5, 3)
print(result)  # Output: 8
```

### 5. **The `re` Module**
The `re` module provides support for regular expressions, allowing you to search, match, and manipulate strings using patterns.

**Common Functions:**
- `re.search(pattern, string)`: Searches for the pattern in the string.
- `re.match(pattern, string)`: Checks for a match only at the beginning of the string.
- `re.findall(pattern, string)`: Returns all matches of the pattern in the string.

**Example:**
```python
import re

pattern = r'\d+'
string = 'There are 123 apples and 456 oranges.'

print(re.findall(pattern, string))  # Output: ['123', '456']
print(re.search(pattern, string).group())  # Output: '123'
```

### 6. **The `time` Module**
The `time` module provides functions for working with time and delays.

**Common Functions:**
- `time.time()`: Returns the current time in seconds since the epoch.
- `time.sleep(seconds)`: Suspends execution for the given number of seconds.
- `time.localtime()`: Returns the local time in a `struct_time` object.

**Example:**
```python
import time

print(time.time())  # Output: e.g., 1674258121.627963
time.sleep(1)       # Sleep for 1 second
print(time.localtime())  # Output: e.g., time.struct_time(tm_year=2024, tm_mon=1, tm_mday=20, ...)
```

````python
import time


print(time.time())
print(time.localtime())
````
