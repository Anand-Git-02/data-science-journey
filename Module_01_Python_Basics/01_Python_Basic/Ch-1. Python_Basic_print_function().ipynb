# Data Types

### Python supports 3 categories of data types:
- Basic Types -(integer,float,complex,boolean and string)
- Container Types - (List,Tuples,Sets and Dictionary)
- User-define Types-Class

```python
# integer
print(6)
print(1e309)
print(type(6))
```

```python
# float
print(4.5)
print(4.1e307)
print(type(4.5))
```

```python
# boolean
print(True)
print(False)
print(type(True))
```

```python
# complex
print(4+5j)
print(type(4+5j))
```

```python
# string
print('hello')
print("hi")
print('''hmm''')
print(type('Hi'))
```

# Container Types

```python
# list

print([1,2,3,4,5,6],end='\n\n')

print(type([1,2,3,4,5,6]))
```

```python
# tuple

print((1,2,3,4,5))
print(type((1,2,3,4,5)))
```

```python
# sets

print({1,2,3,4,5})
```

```python
# Dict

print({"Name":"Anand","Age":22,"gender":"Male"})
```
