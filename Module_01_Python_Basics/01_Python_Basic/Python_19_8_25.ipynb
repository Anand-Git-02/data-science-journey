```python
import time

L = list(range(1000000))
T = tuple(range(1000000))

start = time.time()

for i in L:
    i * 5
print(f"List time  :- {time.time()-start}")
 
start = time.time()
for i in T:
    i * 5
print(f"Tuple time :- {time.time()-start}")
```
