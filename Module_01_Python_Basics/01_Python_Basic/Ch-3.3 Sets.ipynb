# Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:
- Unordered
- Mutable
- No Duplicates
- Can't contain mutable data types

### Set Operation

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}


# Union(|)
print(s1 | s2)

# Intersection(&)
print(s1 & s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
# Difference(-)  The elements not present s1
print(s1 - s2)
print(s2 - s1)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
# Symmetric Difference(^)
s1 ^ s2
```

```python
s1 = {1,2,3,4,5}


print(4 in s1)
# Membership Test
1 not in s1
```

```python
s1 = {1,2,3,4,5}

# Iteration

for i in s1:
    print(i)
```

```python
# union/update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# s1 | s2
s3 = s1.union(s2)
print(s3)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print("S1 :-",s1)
s1.update(s2)
print("S1 after update :-",s1)
print(s2)
```

```python
# intersection/intersection_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# s1 & s2

print(s1.intersection(s2))
print(s1)
s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# difference/difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.difference(s2)
```

```python
s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# symmetric_difference/symmetric_difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.symmetric_difference(s2)

s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint/issubset/issuperset
# Return True if no items in set s1 is present in set s2:
# The isdisjoint() method returns True if none of the items are present in both sets, otherwise it returns False.


s1 = {1,2,3,4}
s2 = {7,8,5,6}

s1.isdisjoint(s2)
```

```python
# The issuperset() method returns True if all items in the specified set exists in the original set, otherwise it returns False.
# As a shortcut, you can use the >= operator instead

s1 = {1,2,3,4,5}
s2 = {3,4,5}
s3 = {4,5}

print(s1.issuperset(s2))

print(s2 >= s2)
```

```python
# copy
s1 = {1,2,3}
s2 = s1.copy()

print(s1)
print(s2)
```

### Frozenset
Frozen set is just an immutable version of a Python set object

```python
# create frozenset
fs1 = frozenset([1,2,3])
fs2 = frozenset([3,4,5])

fs1 | fs2
```

```python
# what works and what does not
# works -> all read functions
# does't work -> write operations
```

```python
# When to use
# 2D sets
fs = frozenset([1,2,frozenset([3,4])])
fs
```
