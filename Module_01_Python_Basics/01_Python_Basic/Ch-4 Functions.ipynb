### What are functions ?

- `Function` is code that has a name and it can be reused(execute again) by specifying its name in the program, whenever needed.

- Function help break our program into smaller parts.

### 1. **Defining a Function**
   - A function is defined using the `def` keyword. It's like setting up a recipe.

```python
def greet():
    print("Hello, Welcome to Python Programming")
    print("You are Learning about functions")
```

### 2. **Calling a Function**
   - After defining a function, you can use it whenever you need by calling its name.


```python
 greet()
```

### Adventages of Function

- Increases readability, particularly for longer code as by using function, the program is better organized and easy to understand.
- Reduces code length as same code is not required to be written at multiple places in a program.This also make debugging easier.

```python
def multiply(*args):
    
    product = 1
    for i in args:
        product  = product * i
        
    return product
```

```python
def multiply(*args):
    
    product = 1
    
    for i in args:
        product  = product * i

    return product

multiply(1,2,4)
```

```python
def multiply(*anyword):
    
    product = 1
    
    for i in anyword:
        product  = product * i

    print(anyword)
    return product

multiply(1,2,4)
```

### 4. **Returning Values**
- Functions can return values after doing some work, just like a recipe can produce a final dish.

```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  
```

### 5. **Using Default Values**
   - You can set default values for parameters, so the function still works even if you don’t provide all the inputs.
   - If you don’t give a name, it defaults to "Guest."

```python
def greet(name="Guest"):
    print(f"Hello, {name}!")

greet()        
greet("Alice")  
```

### 6. **Flexible Inputs with `*args` and `**kwargs`**
   - You can make a function flexible enough to handle any number of inputs.

   - The `*names` allows the function to take any number of names.

```python
def greet(*names):
    for name in names:
        print(f"Hello, {name}!")

greet("Alice", "Bob", "Charlie")
```

```python
def g(x):
    def h(x):
        x = x+1
        print("in h(x): x = ", x)
    x = x + 1
    print('in g(x): x = ', x)
    h(x)
    return x

x = 3
z = g(x)
print('in main program scope: x = ', x)
print('in main program scope: z = ', z)
```

```python
def g(x):
    def h(x):
        x = x+1
        print("in h(x): x = ", x)
        return x
        
    x = x + 1
    print('in g(x): x = ', x)
    h_function = h(x)
    return x , h_function

x = 3
z,h = g(x)
print('in main program scope: x = ', x)
print('in main program scope: z = ', z)
print('Return Value of h(x) =',h)
```

### Higher Order Functions

```python
# Example

def square(x):
    return x**2

def cube(x):
    return x**3

# HOF
def transform(f,L):
    output = []
    for i in L:
        output.append(f(i))

    print(output)

L = [1,2,3,4,5]

transform(square,L)
```

```python
transform(lambda x:x**3,L)
```

```python
# fetch names from a list of dict

users = [
    {
        'name':'Rahul',
        'age':45,
        'gender':'male'
    },
    {
        'name':'Nitish',
        'age':33,
        'gender':'male'
    },
    {
        'name':'Ankita',
        'age':50,
        'gender':'female'
    }
]

list(map(lambda users:users['age'],users))
```
