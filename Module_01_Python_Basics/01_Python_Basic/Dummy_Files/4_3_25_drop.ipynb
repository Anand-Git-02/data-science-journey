```python
text = "Exploring new technologies while honing skills in data science and machine learning to build innovative solutions"


words = text.split()
List_sentences = []

for i in range(0, len(words), 6):
    List_sentences.append(" ".join(words[i:i+6]))

print(List_sentences)
```

```python
with open('Example.txt','w') as file:

    for i in List_sentences:
        file.write(i)
        file.write('\n')
```

```python
n = int(input("How many Number You Want?"))

List = []

for i in range(1,n+1):
        try:
            User_num = int(input(f"Enter Your Number {i}"))
            List.append(User_num)

        except:
            print("String is given")
    
print(List)
```
