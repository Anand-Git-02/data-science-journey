```python
def parse_revenue(text):
    if not pd.isna(text):
        text = text.replace('(USD)','')
        text = text.replace('$','')
    else:
        return np.nan
    return text
```

```python
import pandas as pd
import numpy as np
import re

# Step 1: Clean revenue string
def clean_revenue(revenue_str):
    if pd.isna(revenue_str):
        return np.nan

    revenue_str = revenue_str.lower()

    if "less than" in revenue_str:
        return "0 - 1M"
    if "1 to 5 million" in revenue_str:
        return "1M - 5M"
    if "5 to 10 million" in revenue_str:
        return "5M - 10M"
    if "10 to 25 million" in revenue_str:
        return "10M - 25M"
    if "25 to 50 million" in revenue_str:
        return "25M - 50M"
    if "50 to 100 million" in revenue_str:
        return "50M - 100M"
    if "100 to 500 million" in revenue_str:
        return "100M - 500M"
    if "500 million to 1 billion" in revenue_str:
        return "500M - 1B"
    if "1 to 2 billion" in revenue_str:
        return "1B - 2B"
    if "2 to 5 billion" in revenue_str:
        return "2B - 5B"
    if "5 to 10 billion" in revenue_str:
        return "5B - 10B"
    if "10+ billion" in revenue_str:
        return "10B+"
    if "-1" in revenue_str:
        return np.nan
    
    return revenue_str
```

```python
new_df['Revenue_clean'] = new_df['Revenue'].apply(clean_revenue)
```

```python
# Step 2: Convert cleaned revenue to min/max/avg
def parse_revenue_range(revenue_str):
    if pd.isna(revenue_str):
        return pd.Series([np.nan, np.nan, np.nan])

    # Remove unwanted symbols
    revenue_str = revenue_str.replace('$', '').replace(',', '').replace('usd', '').strip()
    revenue_str = revenue_str.replace('B', 'e9').replace('M', 'e6').replace('+', '')

    if " - " in revenue_str:
        min_val, max_val = revenue_str.split(" - ")
        min_val = float(eval(min_val.strip()))
        max_val = float(eval(max_val.strip()))
        avg_val = (min_val + max_val) / 2
        return pd.Series([min_val, max_val, avg_val])
    
    elif 'e' in revenue_str:  # like "10e9"
        val = float(eval(revenue_str))
        return pd.Series([val, val, val])

    return pd.Series([np.nan, np.nan, np.nan])
```

```python
new_df[["Revenue Min", "Revenue Max", "Revenue Avg"]] = new_df["Revenue_clean"].apply(parse_revenue_range)
```

```python
def extract_education(info):
    education_levels = ['Bachelor', 'Master', 'PhD', 'Doctorate']
    level = []
    for i in education_levels:
        if i.lower() in info.lower():
            level.append(i)
    return list(set(level))

new_df['EducationLevel'] = new_df['Job Description'].apply(extract_education)
```

```python
# years of experience from job description
def extract_experience(description):
    if pd.isna(description):
        return np.nan
    # Pattern matches '3 years', '3+ years', '3-5 years', '3 to 5 years', etc.
    patterns = [
        r'(\d+)\+?\s*(?:year|yr)s?',                      # 3 years, 5yrs, 3+ years
        r'(\d+)\s*[\-to]+\s*(\d+)\s*(?:year|yr)s?'        # 3-5 years, 3 to 5 years
    ]
    
    max_years = []
    for pattern in patterns:
        matches = re.findall(pattern, description, re.IGNORECASE)
        for match in matches:
            if isinstance(match, tuple):
                max_years.extend([int(m) for m in match])
            else:
                max_years.append(int(match))
    
    return max(max_years) if max_years else np.nan
```

```python
new_df['YearsExperience'] =  new_df['Job Description'].apply(extract_experience)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

#  Rating column
rating = new_df['Rating']

# Plot histogram and boxplot
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.histplot(rating, bins=20, kde=True)
plt.title('Histogram of Ratings')

plt.subplot(1, 2, 2)
sns.boxplot(x=rating)
plt.title('Boxplot of Ratings')

plt.tight_layout()
plt.show()

# Compute statistics
mean_rating = rating.mean()
median_rating = rating.median()
mode_rating = rating.mode().iloc[0] if not rating.mode().empty else None
skewness = skew(rating)
kurt = kurtosis(rating)

print(f"Mean: {mean_rating}")
print(f"Median: {median_rating}")
print(f"Mode: {mode_rating}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
```

```python
#### 5. **Which `Company Name`s appear most often?**
   # – List the top 10 companies by posting count, then visualize as a bar chart (or Pareto chart) to see where your data is concentrated.


# Cumulative % for Pareto chart
top_companies = new_df['Company Name'].value_counts().head(10)
cumulative_percentage = top_companies.cumsum() / top_companies.sum() * 100

fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart
sns.barplot(x=top_companies.index, y=top_companies.values, ax=ax1, color='steelblue')
ax1.set_ylabel('Job Postings Count')
ax1.set_xlabel('Company Name')
# ax1.tick_params(axis='x',pad=0.5,width=2,rotation=65)
plt.xticks(rotation=45, ha='right') 
ax1.set_title('Pareto Chart of Top 10 Companies')

# Line chart on secondary axis
ax2 = ax1.twinx()
ax2.plot(top_companies.index, cumulative_percentage[::-1], color='red', marker='o', linewidth=2,)
ax2.set_ylabel('Cumulative Percentage (%)')
ax2.axhline(80, color='gray', linestyle='--', linewidth=1,)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Question 2. *Which `Location's` tend to have higher or lower average `Rating's`?*


```python
# 2. Which Locations tend to have higher or lower average Ratings?
# Group by Location and calculate the mean or median rating.
# Visualize the top 10 locations by listing volume and their average rating with a bar chart.
# Question: Are certain cities associated with better-rated job posts?

# Group by Location and calculate count (number of listings) and average rating
location_stats = new_df.groupby('Location').agg(
    Count=('Location', 'count'),
    Mean_Rating=('Rating', 'mean')
).reset_index()

# Get top 10 locations by listing volume
top_locations = location_stats.sort_values(by='Count', ascending=False).head(10)

# Sort top locations by mean rating for visualization
top_locations = top_locations.sort_values(by='Mean_Rating', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Location', y='Mean_Rating', data=top_locations, palette='cubehelix')
plt.title('Average Rating of Job Posts by Top 10 Locations (by Listing Volume)')
plt.xlabel('Location')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

3. Locations like Chantilly, VA, Chicago, IL, and "United States" have lower average ratings compared to the leaders.
4. Overall, the difference between the highest and lowest average ratings in the top 10 is moderate, suggesting generally consistent candidate experiences across these locations.

- INTERPRETATIONS.

1. Boston’s high rating might reflect better job quality, work culture, or competitive opportunities in that area.
2. McLean and Washington, DC could be high in ratings due to being tech hubs or offering competitive packages and reputations.

### Question 3. **Is there a relationship between Company Name and Rating?**


```python
### 3. **Is there a relationship between Company Name and Rating?**

# * Focus on top 10–15 companies by frequency.
# * Plot *average rating per company*.
# * Question: Do some companies consistently receive higher or lower ratings?

# Group by Company Name and calculate count (number of listings) and average rating
company_stats = new_df.groupby('Company Name').agg(
    Count=('Company Name', 'count'),
    Mean_Rating=('Rating', 'mean')
).reset_index()

# Get top 10 companies by listing volume
top_companies = company_stats.sort_values(by='Count', ascending=False).head(10)

# Sort top companies by mean rating for visualization
top_companies = top_companies.sort_values(by='Mean_Rating', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Company Name', y='Mean_Rating', data=top_companies, palette='Set1')
plt.title('Average Rating of Job Posts by Top Companies (by Listing Volume)')
plt.xlabel('Company Name')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

3. The difference in average ratings across companies is more pronounced than in locations, indicating companies have a stronger effect on job seeker perception.
4. The top-rated companies often have fewer outliers in their ratings, suggesting a more consistent quality of postings.

- INTERPRETATIONS:

1. Companies with higher average ratings may offer better work environments, clearer job roles, or better employer reputation.
2. Lower ratings in some high-volume companies could suggest candidate dissatisfaction due to issues like unclear expectations, poor work-life balance, or other negative factors.
3. The wider variation in company-based ratings highlights the strong role of company-specific factors (policies, culture, benefits) in shaping job seeker experiences.
4. For job seekers, company reputation may be more indicative of rating and satisfaction than just location.



### Question 4. **Does the length of the Job Description relate to the Rating?**


```python
### 4. **Does the length of the Job Description relate to the Rating?**

# * Compute job description length (word or character count).
# * Plot a *scatterplot*: x = description length, y = rating.
# * Add a trendline (e.g., using regression).
# * Question: Do longer or more detailed descriptions tend to correlate with better ratings?

#  Compute description length in words (or use len() for characters)
new_df['Description Length'] = new_df['Job Description'].str.split().str.len()

# Drop rows with missing ratings or job descriptions
plot_data = new_df[['Description Length', 'Rating']].dropna()

# Step 3: Create scatterplot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Description Length', y='Rating', data=plot_data, scatter_kws={'alpha': 0.5})
plt.title('Relationship Between Job Description Length and Rating')
plt.xlabel('Job Description Length (words)')
plt.ylabel('Company Rating')
plt.tight_layout()
plt.show()
```

### Question No : 4

- OBSERVATIONS:

1. The regression line in the scatterplot is nearly flat, indicating no strong upward or downward trend.
2. Most data points are clustered between 3.5 and 4.5 ratings, regardless of description length.
3. Description lengths vary widely, ranging from very short (under 100 words) to very long (over 1000 words).
4. There are several outlier job descriptions with lengths exceeding 1000 words, but they do not correspond to extreme ratings.

- INTERPRETATIONS:

1. The flat trendline suggests that description length has little to no effect on how a company is rated.
2. Since ratings are tightly clustered, candidates likely value other factors (e.g., benefits, culture) more than how detailed a listing is.
3. The wide spread of ratings for mid-length descriptions (300–700 words) shows that employers with both high and low ratings use similarly sized descriptions.
4. Long descriptions may reflect complexity or regulatory requirements, but they don't seem to influence overall company perception or rating.

### Question 5. **Do highly rated jobs tend to be in certain Job Title and Location combinations?**


```python
# 5. Do highly rated jobs tend to be in certain Job Title and Location combinations?
# Create a heatmap or grouped bar chart showing the average rating for combinations of top job titles and locations.
# Question: Is there a “sweet spot” where both job type and location yield higher-rated listings?

# Group by Job Title and Location and calculate average rating
title_location_stats = new_df.groupby(['Job Title', 'Location']).agg(
    Mean_Rating=('Rating', 'mean'),
    Count=('Rating', 'count')
).reset_index()

# Select top Job Titles and Locations by frequency
top_job_titles = new_df['Job Title'].value_counts().head(10).index
top_locations = new_df['Location'].value_counts().head(10).index

# Filter for only these top combinations
filtered_stats = title_location_stats[
    (title_location_stats['Job Title'].isin(top_job_titles)) &
    (title_location_stats['Location'].isin(top_locations))
]

# Pivot table for heatmap
heatmap_data = filtered_stats.pivot(index='Job Title', columns='Location', values='Mean_Rating')

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='gist_stern_r', linewidths=.5)
plt.title('Average Rating for Top Job Titles and Locations')
plt.xlabel('Location')
plt.ylabel('Job Title')
plt.show()
```

#### Question No : 5

- OBSERVATIONS:
 
1. Certain Job Title and Location combinations show significantly higher average ratings, revealing a "sweet spot" where both factors align.
2. Some combinations (e.g., Boston, MA + Data Scientist) are top-rated, indicating regional preference or stronger opportunities in that field.
3. Several locations that ranked average in overall ratings may perform better for specific job titles, showing the importance of context.
4. Some popular job titles consistently maintain good ratings across locations, suggesting role-specific consistency.

- INTERPRETATIONS:

1. Job seekers looking for highly rated jobs should consider both location and title, as these combinations influence satisfaction.
2. High ratings in certain locations for specific titles may point to regional hubs for particular industries (e.g., tech in San Francisco, consulting in Washington DC).
3. Companies may focus on optimizing certain roles in specific locations to maximize candidate satisfaction and performance.
4. This analysis can help job seekers strategically target roles and locations for a better work experience.
