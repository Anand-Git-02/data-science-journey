```python
def correlation(X,Y):
    X = df_num[X].values
    Y = df_num[Y].values

    X_bar = np.mean(X)
    Y_bar = np.mean(Y)


    Numitor = []
    d1 = []
    d2 = []

    for i in range(len(X)):
        a = X[i] - X_bar
        b = Y[i] - Y_bar
        Numitor.append(a*b)

    for i in range(len(Y)):
        a = (X[i] - X_bar) ** 2
        b = (Y[i] - Y_bar) ** 2
        d1.append(a)
        d2.append(b)
    answer = sum(d1) * sum(d2)
    return sum(Numitor)/(answer ** 0.5)

correlation('resid_area','air_qual')
```
