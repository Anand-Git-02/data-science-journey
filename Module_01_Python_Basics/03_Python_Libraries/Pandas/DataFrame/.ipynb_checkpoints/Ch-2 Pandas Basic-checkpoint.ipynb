# access in column with loc and iloc

### **🔹 `df.loc[]` (Label-Based Indexing)**  
- Used for selecting data **by labels** (row/column names).  
- Supports **single labels, lists, slices, and boolean indexing**.  
- When slicing, **both start and end indexes are included**.  

### **🔹 `df.iloc[]` (Position-Based Indexing)**  
- Used for selecting data **by integer positions** (0-based index).  
- Supports **single positions, lists, and slices**.  
- When slicing, **end index is excluded** (like Python lists).  

### **🔹 Key Differences**  
| Feature  | `df.loc[]` | `df.iloc[]` |
|----------|-----------|-------------|
| Indexing Type | Label-based | Position-based |
| Row Selection | By label | By position |
| Column Selection | By column name | By position |
| Slicing | Includes end index | Excludes end index |
| Boolean Masking | Supported | Not supported |
