```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({"fruit" : ["apple", "banana", "orange","grape"],
                    "market_price" : [100, 40, 80, 60]})
df1
```

```python
df2 = pd.DataFrame({"fruit" : ["apple", "banana", "grape",'kiwi'],
                    "wholesaler_price" : [60, 20, 45,100]})
df2
```

```python
# Marge two dataset with how 
# --> Method               Description 
#      left           Use keys from left object 
#      right          Use keys from right object 
#      outer          Use union of keys 
#      inner          Use intersection of keys
```

```python
pd.merge(df1,df2,on='fruit')
```

```python
pd.merge(df1,df2,on='fruit',how='outer')
```

```python
pd.merge(df1,df2,on='fruit',how='left')
```

```python
pd.merge(df1,df2,on='fruit',how='right')
```

```python
pd.merge(df1,df2,on='fruit',how='right',indicator=True)
```

```python
df1 = pd.DataFrame({"fruit" : ["apple", "banana","orange", "grape"],
                    "Vagitables" : ['cucumber','tometo','poteto','carrot'],
                    "market_price" : [100, 40, 80, 60]})
df1
```

```python
df2 = pd.DataFrame({"fruit" : ["apple", "banana", "grape",'dragon_fruit'],
                    "Vagitables" : ['cucumber','tometo','cabbige','carrot'],
                    "wholesaler_price" : [60, 20, 45,100]})
df2
```

```python
pd.merge(df1,df2,on=['fruit','Vagitables'],how='outer')
```

```python
pd.merge(df1,df2,on=['fruit','Vagitables'],how='outer')
```

```python
pd.merge(df1,df2,on='fruit',how='left')
```

```python
df3 = pd.DataFrame({"ID" : [1,2,3,4],
                    "Name" : ["a","b","c","d"]})
df3
```

```python
df4 = pd.DataFrame({"ID" : [5,6,7,8],
                    "Name" : ["a","b","c","d"]})
df4
```

```python
pd.merge(df3,df4,left_index=True,right_index=True)
```

```python
pd.merge(df3,df4,left_index=True,right_index=True,suffixes=['_1','_2'])
```
