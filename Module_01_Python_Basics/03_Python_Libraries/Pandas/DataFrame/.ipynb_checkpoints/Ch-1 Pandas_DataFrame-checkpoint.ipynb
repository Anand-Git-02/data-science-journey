2. **dtypes**  :- Return the data types of data in DataFrame. 
3. **index**   :- The index (row lables) of the DataFrame.
4. **columns** :- The column lables of the DataFrame.
5. **values**  :- Return a Numpy representation of the DataFrame.
6. **axes**    :- Return a list representing both the axes (axis 0 i.e, index and axis 1, i.e, columns).
7. **size**    :- Return an int representing the number of elements in this object.

## If we see above code its only showing "Action" genres but there one problem in our dataframe there 'Action|Drama|War' our code not consider this to solve this problem we use this code

```python
# see the error
movies['genres'].split('|')
```

## That why we use "movies.str.split('l')"

```python
movies['genres'].str
```

```python
movies['genres'].str.split('|')
```

```python
#  Above code will give us 'genres' column in list
# now we applying "Lambda" function

movies['genres'].str.split('|').apply(lambda x : "Action" in x)
```

```python
genres = movies['genres'].str.split('|').apply(lambda x : "Action" in x)
rating = movies['imdb_rating'] > 7.5
```

```python
# type(genres)
# type(rating)
```

```python
movies.head()
```

```python
movies[genres & rating]['genres']
```

```python
movies[genres & rating].shape[0]
```

```python
movies[genres & rating][['original_title','year_of_release','imdb_rating']].sort_values('imdb_rating',ascending=False)
```

## Without using lambda function 

```python
genres = movies['genres'].str.contains('Action')
rating = movies['imdb_rating'] > 7.5

movies[genres & rating].shape[0]
```
