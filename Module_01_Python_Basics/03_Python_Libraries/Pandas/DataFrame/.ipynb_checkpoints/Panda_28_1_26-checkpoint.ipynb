#### 2. How many matches "Super over" finishes occured

```python
ipl['SuperOver'].value_counts()
```

```python
ipl[ipl['SuperOver'] == "Y"][["Team1","Team2","WinningTeam"]]
```

#### 3. How many matches has csk won in Kolkata

```python
# winning team --> Chennai Super Kings
# city --> Kolkata
# and
```

```python
ipl['City'].value_counts()
```

```python
ipl['WinningTeam'] == "Chennai Super Kings"
```

```python
ipl[(ipl['City'] == "Kolkata") & (ipl['WinningTeam'] == "Chennai Super Kings")][["Team1","Team2","City","WinningTeam"]].shape
```

```python
city = (ipl['City'] == "Kolkata")
win_team =  (ipl['WinningTeam'] == "Chennai Super Kings")

ipl[city & win_team][["Team1","Team2","City","WinningTeam"]]
```

```python
ipl[(ipl['City'] == "Ahmedabad") & (ipl['WinningTeam'] == "Chennai Super Kings")][["Team1","Team2","City","WinningTeam"]]
```

#### 5. Moives with rating higher than 8 and votes less 1000

```python
movies.head()
```

```python
rating = movies['imdb_rating'] > 8
votes = movies['imdb_votes'] < 1000

movies[rating & votes][['original_title','year_of_release','runtime','genres','imdb_rating','imdb_votes','actors']]
```
