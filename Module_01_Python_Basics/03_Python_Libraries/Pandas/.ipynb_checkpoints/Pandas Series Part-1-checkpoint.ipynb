# **Pandas Series Data Structure**

### 1. Series is an important data struture of pandas  
### 2. It resprest a one-dimensional array of indexed data

# **A series type object has two main component:**

### ~> an array of **actual data**
### ~> an associated array of **indexes or data labels**.

```python
import pandas as pd
import numpy as np
```

```python
# how to create empty series objects
# <Series objecti> = pandas.Series()
import pandas as pd
obj1 = pd.Series()
obj1
```

```python
# how to create non-empty Series
# <Series objecti> = pd.Series(data,index=idx)

obj2 = pd.Series(range(1,5)) # Using range(5) generate a sequence [0,1,2,3,4]

obj3 = pd.Series([1,43,5.5,66]) # Using list [] 
```

```python
print("Using range function"'\n\n'"INDEX range function")
obj2
# on below output 
# Left side is INDEX of data
# Right side is ACTUAL DATA 
```

```python
print("Using List []""\n\n""INDEX List []")
obj3
```

# **(i). Specify Data as Python Sequence**

```python
# 1. Write code to create Series object using the python sequence[4,6,8,10]

s1 = pd.Series([4,6,8,10])
print("Series object 1:")
print(s1) 
```

```python
# 2. Write code to create Series object using the python sequence(11,21,31,41)

s2 = pd.Series((11,21,31,41))
print("Series object 2:")
print(s2)
```

```python
# 3. Write code to create Series object using individual character "o","h" and "o"

s3 = pd.Series(['o','h','o'])
print("Series object:")
print(s3)
```

```python
# 4. Write code to create Series object using a string : "So funny"

s4 = pd.Series("SO funny")
print("Series object:")
print(s4)
```

```python
# 5. Write code to create Series object using three differt words: "I","am","laughing".

s5 = pd.Series(["I","am","laughing"])
print("Series Object:")
print(s5)
```

# **(ii). Specify Data as an ndarry**

```python
# The data attribut can be ndarray also.
nda1=np.arange(3,13,3.5)
print(nda1)
```

```python
ser1 =  pd.Series(nda1)
print(ser1)
```

```python
# we can combine above code line like this:
ser1 = pd.Series(np.arange(3,13,3.5))
print(ser1)
```

```python
# 6. Write code to create Series object using an ndarray that has 5 element in range 24 to 64 
s6 = pd.Series(np.linspace(24,64,5)) # using np.linspace(starting point,end point,number of element)
print(s6)
```

```python
# Write code to create Series object using an ndarray that is created by tiling a list [3,5], twice.
s7 = pd.Series(np.tile([3,5],2))
print(s7)
```

# **(iii). Specify Data as a Python Dictionary**
## ~> IF we are Create a series object from a dicitionary object,then **Keys** of the Dictionary becomes **index** of the Series and the **Values** of dictionary become the data of Series object 

## ~> Also, the **indexes,which are created from keys may not be in the same order as you have typed them**

```python
# How you can Create a Series object by specifying indexes and values through a dicitionart

obj5 = pd.Series({"Jan":31,"Feb":28,"Mar":31})


print("See, Left column took the keys of passed dicitionary"'\n'"*AS INDEXES*""\n""and values from the""\n""*data values* :"'\n')

obj5
```

```python
# 8. Write code to create Series object using a dictionary that stores 
#    the number of student in each section of class 12 in your school
stu = {"A":39,"B":41,"C":42,"D":44}
s8 = pd.Series(stu)
print(s8)
```

# **(iv). Specify data as a Scalar Value**
## The **data** can be in form of a single value or scalar value.BUT if data is a scalar value, then the **index** argumenat to **Series()** function must be provided.

## ~> The scalar value (given as data) will be repeated to match the length of index

## ~> (The **index** argument has to be a sequence of number or lables of any type)

## for example consider this:


```python
medalsWon = pd.Series(10,index=range(0,1))
print(medalsWon)
```

```python
medals2 = pd.Series(15,index=range(0,6,2)) # range(starting point,end point,skiping number)
print(medals2)
```

```python
ser2 = pd.Series("Yet to start",index = ["Indore","Gujarat","Shimla"])
print(ser2)
```

```python
# 9. Write code to create Series object that stores 
#the initial budget allocated(50000/- each) for the four quarters of the year:Qtr1,Qtr2,Qtr3 and Qtr4

s9 = pd.Series(50000,index=["Qtr1","Qtr2","Qtr3","Qtr4"])

print(s9)
```

```python
# 10. Total number of medals to be won is 200 in the Inter University games held every alternate year.
#  Write code to create Series object that stores these medals for games to be held in the decade 2020-2029.

s10 = pd.Series(200,index=range(2020,2029,2))
print(s10)
```
