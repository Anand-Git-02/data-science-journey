```python
Series_1 = pd.read_csv('Data/subs.csv').squeeze("columns")
Series_2 = pd.read_csv('Data/bollywood.csv',index_col='movie').squeeze('columns')
Series_3 = pd.read_csv('Data/kohli_ipl.csv',index_col='match_no').squeeze('columns')
```

```python
# Series Attributes
# 1. Size
# 2. dtype
# 3. name
# 4. is_unique
# 5. index
# 6. values
```

### 1. Size

```python
print("Series_1 :-",Series_1.size)
print("Series_2 :-",Series_2.size)
print("Series_3 :-",Series_3.size)
```

### 2. dtype 

```python
print("Series_1 :-",Series_1.dtype)
print("Series_2 :-",Series_2.dtype)
print("Series_3 :-",Series_3.dtype)
```

### 3. Name

```python
print("Series_1 :-",Series_1.name)
print("Series_2 :-",Series_2.name)
print("Series_3 :-",Series_3.name)
```

### 4 . is_unique

```python
print("Series_1 :-",Series_1.is_unique)
print("Series_2 :-",Series_2.is_unique)
print("Series_3 :-",Series_3.is_unique)
```

```python
print("Series_1 :-",Series_1.index)
print("Series_2 :-",Series_2.index)
print("Series_3 :-",Series_3.index)
```

### values

```python
print("Series_1 :-",Series_1.values)
print("Series_2 :-",Series_2.values)
print("Series_3 :-",Series_3.values)
```
