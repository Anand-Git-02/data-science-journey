```python
import numpy as np
import pandas as pd
```

### How to Create Series

###### 1. Create empty series objects

```python
S1 = pd.Series([])
type(S1)
```

###### 2. Create non-empty Series Using

```python
S2 = pd.Series([1,2,3,4])
S2
```

###### 3. Create Series from list[ ]

```python
List_1 = [1,2.5,"Goku",True]

S3 = pd.Series(List_1)
S3
```

###### 4. Create Series from Tuples ()

```python
Tuples_1 = (1,2.5,"Goku",True)

S3 = pd.Series(Tuples_1,index=[i for i in 'abcd'])
S3
```

###### 5. Create Series from dict

```python
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

dict_1 = dict(zip(subjects,marks))

dict_1
```

```python
s5 = pd.Series(data=dict_1)
s5
```

```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

s6 = pd.Series(marks,name='Goku ke Marks')
s6
```

###### 6. Create Series from An ndarray

```python
pd.Series(np.array([1,2,3,4,5]))
```
