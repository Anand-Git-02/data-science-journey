# 1.4.2 Creating Series Object - Additional Functionality

#### (i) Specifying/Adding NaN values in a Series Object

- Sometimes we need to create a series object of a certain size but we do not have complete data available at that time  
- In such cases,we can fill missing data with a **NaN** (**Not a Number**) value.

⚛ Legal empty value **NaN** is defined in **Numpy module** and hence we can use **np.NaN** to specify a missing value, or use **None**
for example : 

````python
# Use np.NaN or None to add missing data.

obj3 = pd.Series([6.5,np.NaN,2.34,None])

print("Use (np.NaN or None) to add missing data.\n")

obj3
````

### **(ii) Specify index(es) as well as data with Series**

- While Creating Series type object is that along with ***Values***, you also provide **indexes**.

- Both values and indexes are sequence.

**The syntax for this is as follows:**

< Series Object > = pandas.Series(data = None, index = None)

#### ✴ Both **data** and **index** have to be sequence ;

➡ None is taken by default,if we skip these parameters.

#### for example ;



````python
arr = [31,28,31,30] # A sequence of numbers

mon = ["Jan","Feb","Mar","Apr"] # A sequence of string

obj3 = pd.Series(index=mon,data = arr) 
# Series object created with both values and Indexes given as sequences
# we could skip Keyword data also from above code line
# obj3 = pd.Series(arr,index=mon)


obj3 # Series object create with above statement
````

````python
# Another Series object create with both values and indexes given as sequences 
obj4 =  pd.Series(data = [31,28,31,30],index = ["Jan","Feb","Mar","Apr"])
obj4
````

````python
# We may use loop for defining index sequence also, e.g.,
# Loop used for specifying indexes
s1 = pd.Series(data = range(1,15,3) ,index = [x for x in 'abcde'])
s1
````

#### Warning
## If specifying indexes explicitly using an index sequence, you must provide indexes equal to the number of values in data array;
Provide too few or too many indices will lead to an error - the **ValueError**.

### (iii). Specify Data Type along with data and index
- We can also specify data type along with data and index with Series( ) 
- **as per following syntax :**
**< Series Object > = pandas.Series(data=None, index=None, dtype=None)**

**For Example :**   

````python
arr = np.array([31,28,31,30])
mon = ['Jan','Feb','Mar','Apr']

#this time dtype provided compare it with obj3 object that is create above without specifying dtype
obj5 = pd.Series(data = arr, index = mon , dtype = np.float64)
obj5
````

### (iv). Using a Mathematical function/Expression to Create Data Array in Series( )
### The Series ( ) allows you to define a function or expression that can calculate values for data sequence.It is done following form : 
**< Series object > = pandas.Series(index = None,data = < function | expression >)**

### for example : 

````python
# Numpy array a create as it support vectorized operation
a = np.arange(9,13)
print("Numpy array : \n")
print(a,"\n")

# Data sequence given as a*2 i.e, every value of a doubled

obj7 = pd.Series(index=a,data=a*2)

print("Series object : \n")
obj7
````

````python
obj8 = pd.Series(index=a,data=a**2)
obj8
````

````python
# The vectorized operations on a Numpy array (eg. a*2 or a**2) will be applied 
# on every elemeant of NumpPy array and stored as data part of Series object.
# Now consider another code :

lst = [9,10,11,12]

obj8 = pd.Series(data = 2*lst)
obj8
````

#### **Notice** this time the expression for **data array** involves a **list** and see how it has impacted **the final data array**
- The above code tries to create data array for Series object **obj8** by giving expression **2*lst** 
-  Where **Lst** is a Python list and as you know a number multiplied with list replicate the list those many times and
- Hence the data array has the same list values replicated twice( 🙊 **2*Lst**).

````python
# Its take to much time to make list of alphabates so here is solution 
import string

# initializing empty list
test_list = []

# using string for filling alphabets
test_list = list(string.ascii_uppercase)

# printing resultant list
print ("List after insertion : " + str(test_list))
````

## Important
- While creating a Series object,when you give index array as a sequence then there is no compulsion for the uniqueness of indexes.
- That is you can have duplicate entries in the index array and Python won't raise any error
## For example : 

````python
val = np.arange(2.75,50,9.75)
val
````

````python
# See, index array has duplicate entries
ob3 = pd.Series(val, index= ['A', 'B', 'C', 'B','A'])

# No error !! Series object successfully converted

print("No error !! Series object successfully converted \n")
print("**NOTE** \n")
print("Indices need not be unique in Pandas series Object.\nThis will only cause an error if/when you perform \nan operation that requires unique indices.\n")
ob3
````
