# **1.4.3 Series object Attributes**
- When you create a Series type object all information related to it is available through attributes.
- You can use these attributes in the following format to get information about the Series object.

**< Series object >.< attributes name >**




### **(a). Retrieving Index Array (index attribut) & Data Array (Values attribute) of Series object**

- You can access the **index array** and data **values array** of existing Series object **obj5** as shown below:

```python
# See <object>.index and <object>.values returned the index array and data values array respectively
# Obj5
obj5 = pd.Series({"Jan":31,"Feb":28,"Mar":31})

# <Series object>.index
obj5.index
```

```python
# <Series object>.values
obj5.values
```

```python
# obj6

obj6 = pd.Series(data = np.arange(11,25,3))
print(obj6.index ,"\n")
print(obj6)
```

```python
# obj7

a = np.arange(9,13)
obj7 = pd.Series(index = a,data = a*2)
print(obj7.index,"\n")
print(obj7)
```

```python
# to understand more attributes we use this <series-objects>
obj2 = pd.Series([3.5,5.,6.5,8.])
obj2
```

```python
obj3 = pd.Series([6.5,np.NaN,2.34])
obj3
```

### **(b). Retrieving Data Type (dtype) and Size of Type (itemsize)**
- To retrieve the data type of individual element of a series object,use `< objectname >.dtype`


```python
# The dtype attribute return the type of data values stored in Series object
obj2.dtype
```

```python
# The type(), however, tell the type of the object passed to it
type(obj2)
```

### **(c). Retrirving Shape**
- The shape of a Series object (shape attribute) tells how big it is, i.e.,
### How many elements it contain including missing or empty values (NaNs).
- Since there is only one axis in Series object,it is shown as **(< n>,)** where **n** is the number of element in the object e.g.,




```python
obj2 = pd.Series([3.5,5.,6.5,8.])
obj3 = pd.Series([6.5,np.NaN,2.34])

print(obj2.shape,obj3.shape)
```

### **(d). Retrieving Dimension (number of axis : ndim attribut),Size (size attribut) and Number of Bytes (nbytes attribute)**
- To know about the dimension (number of axis), use 
**< objectname>.ndim**.
- To know about the number of element in the Series object,use
**< objectname>.size**.
- To know Total number of bytes taken by Series object data,use  
**< objectname>**.nbytes

```python
# Series object is 1-dimensional object
obj2 = pd.Series([3.5,5.,6.5,8.])

obj2.ndim
```

```python
# obj2 has 4 element and obj3 has 3 elemenat
obj2 = pd.Series([3.5,5.,6.5,8.])
obj3 = pd.Series([6.5,np.NaN,2.34])

print(obj2.size,obj3.size)
```

```python
# obj2 has 4 elemeant, hence 4*8 = 32 bytes and 
# obj3 has 3 elements, hence 3*8 = 24 bytes

obj2 = pd.Series([3.5,5.,6.5,8.])
obj3 = pd.Series([6.5,np.NaN,2.34])

print(obj2.nbytes,obj3.nbytes)
```

### **(e). Checking Emptiness (empty attribute) and Presence of NaN (hasnans attribut)**
- We have create am empty Series object namely **obj1** to check emptiness use,
**< objectname>.empty**
- Similarly, to check if a Series object contains some NaN values or not, you can use **hasnans** attribute

### **Note**
⚛ if you use **len( )** on Series object, then it returns total element in it including **NaNs**

⚛ But **< series>.count( )** returns only the count of non-NaN values in a Series object 

```python
# The object obj1 is empty but obj2 and obj3 are not empty as they contain some data
obj1= pd.Series()
obj1.empty
```

```python
# obj2 and obj3
obj2 = pd.Series([3.5,5.,6.5,8.])
obj3 = pd.Series([6.5,np.NaN,2.34])

print(obj2.empty,obj3.empty)
```

```python
# hasnans attribute tells the presence of NaN values

obj2 = pd.Series([3.5,5.,6.5,8.])
obj3 = pd.Series([6.5,np.NaN,2.34])

print("The presence of NaN values in obj2 "":",obj2.hasnans)

print("The presence of NaN values in obj3 "":",obj3.hasnans)
```

```python
# Count() returns count of non_NaN values and len() gives total no.of elements
# obj2
print("The total number of element in obj2 :",len(obj2))
print("The total number of non_NaN elements in obj2 :",obj2.count(),"\n")
# obj3
print("The total number of element in obj3 :",len(obj3))
print("The total number of non_NaN elements in obj3 :",obj3.count())
```

```python
obj5
```

```python
# Slice object with values reversed

obj7[: : -1]
```

## 1.4.5 Operation on Series Object

### 1. Modifying Elements of Series Object
The data values of a series object can be easily modified through item assignment, i.e.,
-  < SeriesObject >[< index >] = < new_data_value >
-  < SeriesObject >[ start : stop ] = < new_data_value >

```python
obj1 = pd.Series([1.50,12.75,24.00,35.25,46.50])
obj1
```

```python
obj1[0] = 1.85
```

```python
obj1
```

```python
obj1[2:4] = -15.75
obj1
```

###  Renaming Indexes
You can even change or rename indexes of a Series object by assifning new index array to its index attributes
- < Object >.index = < new index array >

```python
obj2 = pd.Series([1.50,12.75,24.00,35.25,46.50],index=["a","b","c","d","e"])
obj2
```

###  N O T E
- Series object's values can be modified but size cannot
- So you can say that Series objects are **value-mutable but size-immutable objects.**

```python
obj2.index = ['v','w','x','y','z']
obj2
```

# 2. The head( ) and tail( ) Functions
1. The **head( )** function is used to fetch frist n rows from a Pandas object and 

2. **tail( )** function returns last n rows from a pandas object.
### The syntax to us these function is:
- **< pandas object >.head([n])**
- **< pandas object >.tail([n])**

3. If you do not provide any value for n then **head( )** and **tail( )** will return **frist 5** and **last 5** rows respectively.

```python
ob7 = pd.Series([23,67,12,78,34,32,39,44,55,100,92])
ob7
```

```python
ob7.head()
```

```python
ob7.head(7)
```

```python
ob7.tail()
```

```python
ob7.tail(7)
```

# 6. Sorting Series Values
You can sort the values of Series object on **the basis of values and indexes.**`

### Sorting On the Basis of Values
1. To sort a Series object on the basis of values,you may use **sort_values()** function 
- **< Series Object >.sort_values([ascending=True|False])**

```python
# The argument ascending is optional and if skipped
# It takes the value "True" by default
```

```python
# Ascending order by default

s11.sort_values()
```

```python
# Descending order

s11.sort_values(ascending=False)
```

### Sorting on the Basis of Indexes
1. To sort a Series object on the basis of Indexes,you may use sort_index() function
- **< Series Object >.sort_index([ascending=True|False])**

```python
# Ascending order by default

s11.sort_index()
```

```python
# Descending order

s11.sort_index(ascending=False)
```
