### What is Pandas

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.

https://pandas.pydata.org/about/index.html

### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### A series type object has two main component:

- an array of **actual data**
- an associated array of **indexes or data labels**.

## How to Create Series

### 1. Create empty series objects
``` <Series object> = pandas.Series() ```

````python
obj1 = pd.Series()
obj1
````

### 2. Create non-empty Series Using
``` <Series objecti> = pd.Series(data,index=idx) ```

````python
data = list(range(1,5))
data
````

````python
obj2 = pd.Series(range(1,5)) # Using range(5) generate a sequence [0,1,2,3,4]

print("Using range function")
# on below output 
# Left side is INDEX of data
# Right side is ACTUAL DATA

print(f"\nINDEX range function")
obj2
````

### 3. Create Series from list[ ]

````python
# string
country = ['India','Pakistan','USA','Nepal','Srilanka']

pd.Series(country)
````

````python
# integers
runs = [13,24,56,78,100]

runs_ser = pd.Series(runs)
runs_ser
````

````python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
````

````python
# setting a name
marks = pd.Series(marks,index=subjects,name='Anand ke marks')
marks
````

### 4. Create Series from Tuples ()

````python
Ages = (23,22,19,34,15,18)

obj3 = pd.Series(Ages)
obj3
````

### 5. Create Series from dict

````python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks,name='Anand ke marks')
marks_series
````

### 6. Create Series from An ndarray 

````python
# The data attribut can be ndarray also.
nda1=np.arange(3,13,2)
print(nda1)
````

````python
ser1 =  pd.Series(nda1)
print(ser1)
````

````python
# we can combine above code line like this:

ser1 = pd.Series(np.arange(3,13,3.5))
print(ser1)
````
