### Functions and Lambda function on Pandas Series

`subs.apply(lambda x:'good day' if x > subs.mean() else 'bad day')`

````python
subs
````

````python
subs.mean()
````

````python
# Mean > X-value --> "Good Day"
# Mean < X-value --> "Bad Day"
````

````python
def day_check(x):
    
    if x > subs.mean():
        return "Good Day"
    else:
        return "Bad Day"
````

````python
subs.apply(day_check)
````

````python
subs.apply(day_check).value_counts()
````

````python
subs.apply(lambda x:"Good Day" if x > subs.mean() else "Bad Day")
````

````python
subs.apply(lambda x:"Good Day" if x > subs.mean() else "Bad Day").value_counts()
````

```python
import random

secret = random.randint(1, 20)
tries = 5

print("🎲 Guess the number (1 to 20). You have 5 tries!")

while tries > 0:
    guess = int(input("Your guess: "))

    if guess == secret:
        print("✅ Boom! Correct.")
        break
    elif guess < secret:
        print("📉 Too low.")
    else:
        print("📈 Too high.")

    tries -= 1
    print("Tries left:", tries)

if tries == 0:
    print("💀 Out of tries. The number was:", secret)
```
