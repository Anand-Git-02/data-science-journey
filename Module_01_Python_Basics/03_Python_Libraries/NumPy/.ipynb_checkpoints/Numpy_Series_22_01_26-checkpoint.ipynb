#### How to Create Series

```python
import pandas as pd
```

```python
List = [1,2,3,4]
Tuple = [2,3,6,8,9]
dict_1 = {
    "a":1,
    "b":2,
    "c":3
}
```

```python
pd.Series()
```

```python
S1 = pd.Series(List)
S1
```

```python
S2 = pd.Series(Tuple)
S2
```

```python
S3 = pd.Series(dict_1)
S3
```

```python
pd.Series(range(1,11,2))
```

```python
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']
```

```python
# 

S4 =pd.Series(data=marks,index=subjects)
S4
```

```python
S5 =pd.Series(data=marks,index=subjects,name="meet ke marks")
S5
```
