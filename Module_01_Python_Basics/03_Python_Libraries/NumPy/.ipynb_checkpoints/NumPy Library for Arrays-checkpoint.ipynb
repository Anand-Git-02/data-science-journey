# 5.1 One-Dimensional Array 

A 1-D array is an array of elements of silmilar data type starting from index 0 and ending at n+1.

### 5.1.1 Creating a 1-D Array  

### Program to create array from numpy library

```python
import numpy as np
```

### 1. Creating an array of integers.

```python
myarray1 = np.array([11,22,33,44,55,66],int)

print("The array of integers is :\n",myarray1)
```

### 2. Creating an array of characters.

```python
myarray2 = np.array(['a','b','c','d','e','f'])
print("The array of characters is:\n",myarray2)
```

### 3. Creating an array of strings.

```python
myarray3 = np.array(["JAVA","PYTHON","C++","C"], dtype=str)

print("The array of string is:\n",myarray3)
```

### 4. Creating a float array.

```python
myarray4 = np.array([6.3,24.5,16.7,28.9,85.8])

print("The array of decimal numbers is :\n",myarray4)
```

### 5. Using viwe() function to create a new array from an existing array.

```python
myarray5 = myarray3.view()

print("Using view function to create a new array:\n",myarray5)
```

### 6. Using copy() function to create a new array from an existing array.

```python
myarray6 = myarray4.copy()

print("Using copy function to create a new array:\n",myarray6)
```

## Program for creating arrays using different functions

```python
import numpy as np
```

### 7. Creating an integer array using arange() function.
 

```python
myarray7 = np.arange(14)
print("The array of integers using arange() function is :\n",myarray7)
```

### 8. Creating a float array using arange() function.

```python
myarray8 = np.arange(14,19, dtype=float)

print("The array of decimal number using arange() function is:\n",myarray8)
```

### 9. Creating an integer array using linspace() for equal spacing.

```python
myarray9 = np.linspace(12,24,7)

print("The array of integers using linspace() function is:\n",myarray9)
```

### 10. Creating float array using linspace() for equal spacing.

```python
myarray10 = np.linspace(1., 4., 5)

print("The array of decimal numbers using linspace() function is:\n", myarray10)
```

### 11. Creating array using Zeros() function

```python
myarray11 = np.zeros((4))

print("An array using Zeros:\n",myarray11)
```

### 12. Creating array using Ones() function.


```python
myarray12 = np.ones((5))

print("An array using Ones:\n",myarray12)
```

```python
print("Multiplying an elements to array elements is:\n")
myarray14 * 2
```

### 14.4  Dividing an element from array elements

```python
print("Dividing an element from array elements is:\n")
myarray14 / 4
```

```python
print("Using modulus  operator for the array elements:\n")
myarray14 % 4
```

## Program to use mathematical operators on multiple arrays. 

```python
import numpy as np
```

### 15-16-17. Creating three integer arrays. 

```python
myarray15 = np.array([11,22,33],int)
myarray16 = np.array([40,50,60],int)
myarray17 = np.array([7,8,9],int)
```

### 18. Performing addition and subtraction on all elements of an array.

```python
myarray18  = myarray17 + myarray16 - myarray15

print("Addition and Subtraction on multiple arrays:")
myarray18
```

### 19. Performing multiplication and division on all elements of an array. 

```python
myarray19 = (myarray15 * myarray16)/myarray17

print("Multiplication and division on multiple arrays:")
myarray19
```

```python
print(myarray16)
print(myarray17)
```

### 20. Using modules operator (%) on all elements of the array.

```python
myarray20 = myarray16 % myarray17

print("Using modules operator on two arrays:")
myarray20
```

### 21. Using integer division (//) on all elements of the array. 

```python
myarray21 = myarray16 // myarray17
print("Using integer division on two arrays:")
myarray21
```
