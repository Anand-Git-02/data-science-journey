```python
class Atm:
    
    # constructor(special function) --> superpower --> this function don't need call explicitly 
    def __init__(self):
        self.pin = ''
        self.balance = 0
        self.menu()

    def menu(self):
        User_input = input("""
        Hi How can I help you?
        1. Press 1 to Create Pin
        2. Press 2 to Change Pin
        3. Press 3 to Check Balance
        4. Press 4 to Withdraw
        5. Anything else to exit
        """)

        if User_input == '1':
            # Creating Pin
            self.creat_Pin()
        elif User_input == '2':
            # Change Pin
            self.change_pin()
        elif User_input == '3':
            # Check Balance
            self.check_balance()
        elif User_input == '4':
            # Withdraw Money
            self.withdraw()
        else:
            exit()

    def creat_Pin(self):
        # Pin Variable
        User_pin = input('Enter Your Pin')
        self.pin = User_pin

        # Balance Vaialbe
        User_Balance = int(input("Enter Balance :-"))
        self.balance = User_Balance

        print("Pin Created Successfilly")
        self.menu()

    def change_pin(self):
        old_pin = input("Enter Your Old Pin :-")

        if old_pin == self.pin:
            new_pin = input('Enter Your New Pin :-')
            self.pin = new_pin
            print('Pin Change Successfilly')
            self.menu()

        else:
            print('Nahi Hoga ree !!')
            self.menu()

    def check_balance(self):
        User_pin = input("Enter Your Pin")

        if User_pin == self.pin:
            print("Your Balance is",self.balance)

        else:
            print("Chal nikal yahan se !!")
        self.menu()

    def withdraw(self):
        User_pin = input("Enter Your Pin")

        if User_pin == self.pin:
            amount = int(input("Enter Your Amount :-"))

            if amount < self.balance:
                self.balance = self.balance - amount
                print(f"Withdrawl Successful... Your Balance is {self.balance}")
                
            else:
                print("Ek kaam Kar Bank loot le ......")
        else:
            print("Chor ho kaya !!")
        self.menu()
            
    
```

### Magic Methods --> A.K.A Dunder Methods
1. The Constructor __ inti __( )
- Magic Methods are Special Methods every Magic method has superpower
- __ name __ , __ init __
- They are Special methods in Python which are define inside class
- They begin and end with double underscore (__ init __( ))
- They have a Special where in they get called themselves.
    - Creation --> `__ init __`
    - Print    --> `__ str __`
    - '+'      --> `__add__`
    - '-'      --> `__sub__`
    - '*'      --> `__mul__`
    - '/'      --> `__truediv__`
    - '//'     --> `__floordiv__`
    - '=='	   --> `__eq__`

#### What is Constructor
- A constructor is a special method defined within a class that has the unique ability to automatically invoke itself whenever an object of the class is created.

#### What is True benefit Constructor
1. **Automatic Initialization**:
   - When `person1 = Person("Alice", 25)` is executed, the `__init__()` method initializes the `name` and `age` attributes for `person1`.

2. **Simplified Object Creation**:
   - The attributes are set during object creation, so you don’t need a separate method to set them manually.

3. **Consistency**:
   - Every `Person` object is guaranteed to have its attributes (`name` and `age`) initialized properly.


```python
class Person:
    # Constructor
    def __init__(self, name, age):
        self.name = name  # Initialize name
        self.age = age    # Initialize age

    # Method to display information
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating objects using the constructor
person1 = Person("Alice", 25)
person2 = Person("Bob", 30)

# Displaying object information
person1.display_info()  # Output: Name: Alice, Age: 25
person2.display_info()  # Output: Name: Bob, Age: 30
```

### Without Constructor

```python
class Person:
    def set_attributes(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

person1 = Person()
person1.set_attributes("Alice", 25)  # Manual initialization
person1.display_info()  # Output: Name: Alice, Age: 25
```
