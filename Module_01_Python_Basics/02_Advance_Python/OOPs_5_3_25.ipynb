```python
class Account:

    def __init__(self,balance,Account_No):
        self.balance = balance
        self.Account_No = Account_No

    def check_balance(self):
        return f"Your Balance is {self.balance}"

    def credit(self,amount):
        self.amount = amount
        self.balance += self.amount
        
        return f"{self.amount} is Credited"

    def debit(self,amount):
        self.amount = amount

        if self.amount < self.balance:
            self.balance = self.balance - self.amount
            return f"{self.amount} is debited"
        else:
            return f"Chor ho kaya"

account_1 =  Account(10000,12345)
account_1.check_balance()
account_1.debit(100)
account_1.check_balance()
account_1.debit(10000)
account_1.credit(111)
account_1.check_balance()
```

```python
class Atm:
    def __init__(self):
        self.pin = '1234'
        self.balance =  10000
        self.menu()

    def menu(self):
        User_Input = int(input(
            """
            Hi How can I help you?
                1. Press 1 to Create Pin
                2. Press 2 to Change Pin
                3. Press 3 to Check Balance
                4. Press 4 to Withdraw
                5. Anything else to exit
            """
        ))

        if User_Input == 1:
            self.Create_Pin()

        elif User_Input == 2:
            self.Change_Pin()

        elif User_Input == 3:
            self.Check_Balance()

        elif User_Input == 4:
            self.Withdraw()

        elif User_Input == 5:
            exit()
        
        else:
            print("Invalid Input")



    # done
    def Create_Pin(self):
        self.pin = input("Enter Your Pin")
        self.balance = int(input("Enter Your Balance"))

        print(f"Your Pin is {self.pin} and Your Balance {self.balance}")

        self.menu()

    
    # done
    def Change_Pin(self):
        old_pin = input("Enter your Old Pin :-")
        
        if self.pin == old_pin:
            new_pin = input("Enter Your New Pin :-")

            self.pin = new_pin
                    
            print('Pin Change Successfilly')
            
        else:
            print('Nahi Hoga ree !!')
            
    # done
    def Check_Balance(self):
        print(f"Your Balance is {self.balance}")

        self.menu()
        

    def Withdraw(self):
        amount = int(input("Enter Your Amount :- "))

        if amount < self.balance:
            self.balance = self.balance - amount
            print(f"{amount} is debited")
        else:
            print(f"Chor ho kaya")
    

atm1 = Atm()
atm1.menu()
```

```python
## Encapsulation With Getter And Setter
class Person:
    def __init__(self,name,age):
        self.__name=name  ## Private access modifier or variable
        self.__age=age ## Private variable

    ## getter method for name
    def get_name(self):
        return self.__name
    
    ## setter method for name
    def set_name(self,name):
        self.__name=name

    # Getter method for age
    def get_age(self):
        return self.__age
    
    # Setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative.")


person1=Person("Tony",34)
person1.get_name()
person1.set_name("Goku")
person1.get_name()
```

```python
class Car:
    @staticmethod
    def start():
        return "Car started.."

    @staticmethod
    def stop():
        return "Car stopped."

class ToyotaCar(Car):
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Toyota Car: {self.name}"

c1 = ToyotaCar("Model S")
c1.name
c1.start()
c1.stop()
```

```python
class Student:
    def __init__(self,phy,chem,math):
        self.phy = phy
        self.chem = chem
        self.math = math
        self.percentage = str((self.phy + self.chem + self.math)/3) + "%"

std1 = Student(98,97,99)
print(std1.percentage)

# If we want to change the marks it will change it but percentage not change

std1.math = 90
print(std1.math)
print(std1.percentage)
```

```python
class Student:
    def __init__(self,phy,chem,math):
        self.phy = phy
        self.chem = chem
        self.math = math

    @property
    def percentage(self):
        return f"{(self.phy + self.chem + self.math) / 3:.2f}%"

# Markrs in Percentages
std1 = Student(98,97,99)

print(std1.chem)
# Finding of Percentage Average
print(std1.percentage)


std1.chem = 85
print(std1.chem)
print(std1.percentage)
```
