### Procedural Approach in Programming
The procedural approach in programming is a paradigm centered around procedures or routines (functions) to perform tasks. It is one of the oldest programming paradigms and is still widely used in many contexts today. This approach is well-suited for solving problems that can be broken down into a sequence of steps or instructions.

```python
# Calculator functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b if b != 0 else "Error: Division by zero"

# Main function
def main():
    operations = {
        1: ("Addition", add),
        2: ("Subtraction", subtract),
        3: ("Multiplication", multiply),
        4: ("Division", divide)
    }

    print("Simple Calculator")
    for key, (name, _) in operations.items():
        print(f"{key}. {name}")

    try:
        choice = int(input("Enter your choice (1-4): "))
        if choice not in operations:
            print("Invalid choice! Please select a number between 1 and 4.")
            return

        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        operation_name, operation_func = operations[choice]
        result = operation_func(num1, num2)

        print(f"The result of {operation_name.lower()} is: {result}")
    except ValueError:
        print("Invalid input! Please enter numeric values.")

# Execute the program
if __name__ == "__main__":
    main()
```

```python
# Function to add two numbers
def add(a, b):
    return a + b

# Function to subtract two numbers
def subtract(a, b):
    return a - b

# Function to multiply two numbers
def multiply(a, b):
    return a * b

# Function to divide two numbers
def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "Error: Division by zero"

# Main program
def main():
    print("Simple Calculator")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    
    # Get user's choice
    choice = int(input("Enter your choice (1-4): "))
    
    # Input numbers
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    # Perform the selected operation
    if choice == 1:
        result = add(num1, num2)
        print(f"The result of addition is: {result}")
    elif choice == 2:
        result = subtract(num1, num2)
        print(f"The result of subtraction is: {result}")
    elif choice == 3:
        result = multiply(num1, num2)
        print(f"The result of multiplication is: {result}")
    elif choice == 4:
        result = divide(num1, num2)
        print(f"The result of division is: {result}")
    else:
        print("Invalid choice! Please select a number between 1 and 4.")

# Execute the main program
main()
```

### What is An Object And What is OOP ?

An **object** is a specific instance of a class in object-oriented programming (OOP). It combines data (called attributes) and actions (called methods) that represent a real-world thing.

**Object-Oriented Programming (OOP)** is a way of writing software that focuses on using objects to design and organize code. It has four main ideas:

1. **Encapsulation**: Keeps data safe by grouping it with the functions that work on it, so other parts of the program can't change it directly.
2. **Inheritance**: Allows one class (child) to use features from another class (parent), making it easier to share code.
3. **Polymorphism**: Lets different classes use the same function name in different ways, providing flexibility.
4. **Abstraction**: Hides complex details and shows only what is necessary to use an object.



### 1. Class 
**A class is a blueprint or template** in object-oriented programming (OOP) that is used to create objects. It defines the attributes (data) and methods (functions) that the objects will possess, `establishing their structure and behavior`. By using a class, you can create multiple instances (objects) with the same characteristics and functionalities, making it easier to model real-world entities and organize code. Classes promote reusability, encapsulation, and modularity, allowing developers to write clearer and more efficient programs.

#### What is True benefit Constructor
1. **Automatic Initialization**:
   - When `person1 = Person("Alice", 25)` is executed, the `__init__()` method initializes the `name` and `age` attributes for `person1`.

2. **Simplified Object Creation**:
   - The attributes are set during object creation, so you don’t need a separate method to set them manually.

3. **Consistency**:
   - Every `Person` object is guaranteed to have its attributes (`name` and `age`) initialized properly.

#### Car
- Data / Attributes / Property

    - Name
    - Color
    - Brand
    - Top Speed
    
- Behaviour / Methos
    - Accelerate()
    - Brake()
    - Start()

```python
class Car:
    def __init__(self,name,brand,color,topSpeed):
        self.name = name
        self.color = color
        self.brand = brand
        self.topSpeed = topSpeed

    
    def Start(self):
        return f"{self.name} is Started"

    def Accelerate(self):
        return f"Your speed is {self.topSpeed//3} kms"

    def Breke(self):
        return f"{self.name} is stopped"
```

```python
car1 = Car("Model S", "Tesla", "Red", 250)
car2 = Car("Civic", "Honda", "Blue", 220)
car3 = Car("Mustang", "Ford", "Black", 240)
car4 = Car("911 Carrera", "Porsche", "White", 290)
car5 = Car("Camry", "Toyota", "Silver", 210)
```

```python
# Displaying car details
cars = [car1, car2, car3, car4, car5]
for i in range(0,len(cars)):
    print(f"Car {i+1} Name :- {cars[i].name,cars[i].brand}")
```

```python
car1.Accelerate()
```
