```python
class Account:
    balance = 0
    account_no = 123456

    def __init__(self,balance=balance,account_no=account_no):
        self.balance = balance
        self.account_no = account_no

    def credit(self,amount):
        self.balance = self.balance + amount
        return f"{amount} is credited\nYour Balance is {self.balance}"

    def debit(self,amount):
        if amount < self:
            self.balance = self.balance + amount
            return f"{amount} is Credited  Your Balance is {self.balance}"
        else:
            return f"{amount} is Debit  Your Balance is {self.balance}"

ac = Account()
print(ac.credit(200))
```

Here’s a clean **class design** with attributes and methods (no code):

---

### 🔷 **Class: OperatingSystem**

**Attributes:**

* `desk_inv` — inventory ID of the system
* `name` — name of the operating system
* `version` — version number (e.g., 10.0, 22.04)
* `architecture` — system architecture (e.g., x86, x64)

**Methods:**

* `system_info()` — displays complete system information
* `is_supported()` — checks if the OS version is still supported
* `update_version(new_version)` — updates the current OS version

---

### 🔷 **Class: Windows (inherits from OperatingSystem)**

**Additional Attribute:**

* `license_key` — product key for activating Windows

**Additional Method:**

* `activate()` — activates Windows using the license key

---

### 🔷 **Class: Linux (inherits from OperatingSystem)**

**Additional Attribute:**

* `distribution` — name of the Linux distribution (e.g., Ubuntu, Fedora)

**Additional Method:**

* `install_package(package_name)` — installs a software package on the system

---

```python
# Class: OperatingSystem
# desk_inv — inventory ID of the system
# name — name of the operating system
# version — version number (e.g., 10.0, 22.04)
# architecture — system architecture (e.g., x86, x64)
# system_info() — displays complete system information
# is_supported() — checks if the OS version is still supported
# update_version(new_version) — updates the current OS version


class OperatingSystem:

    def __init__(self,desk_inv,name,version,architecture):
        self.desk_inv = desk_inv
        self.name = name
        self.version = version
        self.architecture = architecture

    def system_info(self):
        print(f"""
                Name         :- {self.name}
                Version      :- {self.version}
                Architecture :- {self.architecture}
                Desk_inv     :- {self.desk_inv}
                """)

    def is_supported(self):
        if self.version > 5.0:
            print('The system is supported...')
        else:
            print('The system is not supported.')
        
    def update_version(self,new_version):
        print(f"Current Version :- {self.version} is Update...")
        self.version = new_version
        print(f'Updated Version :- {self.version}')
        
os1 = OperatingSystem("INV-2025-001", "Windows", 10.0, "x64")
os2 = OperatingSystem("INV-2025-002", "Linux", 22.04, "x64")
os3 = OperatingSystem("INV-2025-003", "macOS", 13.3, "ARM64")
os4 = OperatingSystem("INV-2025-004", "Ubuntu", 20.04, "x86")
os5 = OperatingSystem("INV-2025-005", "Fedora", 38, "x64")

os1.system_info()
os1.is_supported()
os1.update_version(24.5)
os1.system_info()
```

```python
# Earth
# soil
# water
# plant


class Earth:
    def __init__(self,name,number):
        self.name =name
        self.number = number
    def Hello_Earth(self):
        print('Inside the Earth Class')

class Soil(Earth):
    def Hello_Soil(self):
        print('Inside the Soil Class')

class Water(Soil):
    def Hello_Water(self):
        print('Inside the Water Class')

class Plant(Water):
    def __init__(self,name,number,plant_number):
        super().__init__(name,number)
        self.plant_number = 2
        
        
    def Hello_Plant(self):
        print('Inside the Plant Class')

plant1 = Plant("Earth",1,1)
plant1.name
```
