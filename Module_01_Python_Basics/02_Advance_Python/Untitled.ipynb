```python
# Calculator functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b if b != 0 else "Error: Division by zero"


def main():
    operations = {
        1: ("Addition", add),
        2: ("Subtraction", subtract),
        3: ("Multiplication", multiply),
        4: ("Division", divide)
    }
    print("Simple Calculator")

    for key, (name, _) in operations.items():
        print(f"{key}. {name}")

    try:
        choice = int(input("Enter your choice (1-4): "))
        if choice not in operations:
            print(f"Invalid choice! Please select a number between 1 and 4.")

        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))

        operation_name, operation_func = operations[choice]
        result = operation_func(num1, num2)

        print(f"The result of {operation_name.lower()} is: {result}")

    except ValueError:
        print("Invalid input! Please enter numeric values.")

if __name__ == "__main__":
    main()
```

```python
class Complex:

    def __init__(self,real,img):
        self.real = real
        self.img = img

    def __str__(self):
        return f"({self.real}+{self.img}j)"

    def __add__(self,other):
        resultReal = 0
        resultImag = 0

        resultReal = self.real + other.real
        resultImag = self.img + other.img

        return f"({resultReal}+{resultImag}j)"

    def __sub__(self,other):
        resultReal = 0
        resultImag = 0

        resultReal = self.real - other.real
        resultImag = self.img - other.img

        if resultImag < 0:
            return f"({resultReal}{resultImag}j)"
        else:
            return f"({resultReal}+{resultImag}j)"

    def __mul__(self,other):
        resultReal = 0
        resultImag = 0

        resultReal = (self.real * other.real) - (self.img * other.img)
        resultImag = (self.real * other.img) + (self.img * other.real)

        return f"({resultReal}+{resultImag}j)"

    def __truediv__(self,other):
        resultReal = 0
        resultImag = 0

        resultReal = (self.real * other.real) + (self.img * other.img)
        resultImag = (self.img * other.real) - (self.real * other.img)

        denometers = (other.real ** 2) + (other.img ** 2)

        if resultImag < 0:
            return f"({resultReal/denometers}{resultImag/denometers}j)"
        else:
            return f"({resultReal/denometers}+{resultImag/denometers}j)"

    def __eq__(self,other):

        if (self.real == other.real) and (self.img == other.img):
            return True
        else:
            return False

        

c1 = Complex(2,3)
c2 = Complex(2,3)
# print(c1)

# print(c1 + c2)
# print(c1 - c2)
# print(c1 * c2)
# print(c1 / c2)
print(c1 == c2)
```

```python
class ComplexNumber:
    def __init__(self,real=0,imag=0):
        self.real = real
        self.imag = imag

    def __str__(self):
        if self.real == 0:
            return f"{self.imag}j"
        elif self.imag < 0:
            s = f"({self.real} {self.imag}j)"
            return s
        else:
            s = f"({self.real} + {self.imag}j)"
            return s
            
    def conjugate(self):
        imag = self.imag * -1
        return f"({self.real}{imag}j)"

    def __add__(self,other):
        resultReal = 0
        resultImag = 0

        resultReal = self.real + other.real
        resultImag = self.imag + other.imag

        ans = ComplexNumber(resultReal,resultImag)

        return ans
        
    def __sub__(self,other):
        resultReal = 0
        resultImag = 0

        resultReal = self.real - other.real
        resultImag = self.imag - other.imag

        ans = ComplexNumber(resultReal,resultImag)

        return ans
    def __mul__(self,other):
        resultReal = 0
        resultImag = 0

        resultReal = (self.real * other.real) - (self.imag * other.imag)
        resultImag = (self.real * other.imag) + (self.imag * other.real)

        ans = ComplexNumber(resultReal,resultImag)

        return ans

    def __truediv__(self,other):
        denominator = other.real ** 2 + other.imag ** 2

        resultReal = 0
        resultImag = 0

        if denominator == 0:
            raise ZeroDivisionError("Cannot divide by zero in complex numbers.")
        
        else:           # (real_1 * real_2) + (imag_1 * imag_2)
                        # (imag_1*real_2)  - (real_1 * imag_2) 
            resultReal = ((self.real * other.real) + (self.imag * other.imag))/denominator
            resultImag = ((self.imag * other.real) - (self.real * other.imag))/denominator

            ans = ComplexNumber(resultReal,resultImag)

            return ans
    def __eq__(self,other):

        if (self.real == other.real) and (self.imag == other.imag):
            return True
        else:
            return False

cn1 = ComplexNumber(2,5)
print(cn1)
print(cn1.conjugate())
```

```python
class Student: #student class
    def __init__(self,name,grade,percentage):
        self.name = name
        self.grade = grade
        self.percentage = percentage

    def student_details(self): #method
        return f"{self.name} is in class {self.grade} with {self.percentage}%"

class GraduateStudent(Student):
    def __init__(self,name,grade,percentage,stream):
        super().__init__(name,grade,percentage)
        self.stream = stream

    def student_details(self): #method
        print(f"{self.name} is in class {self.grade} with {self.percentage}%")
        return f"This is Same Method but different Class"


s1 = Student("Jadu",11,96)

Gs1 = GraduateStudent("Raju",12,95,"PCM")


print(s1.student_details())
print(Gs1.student_details())
```
