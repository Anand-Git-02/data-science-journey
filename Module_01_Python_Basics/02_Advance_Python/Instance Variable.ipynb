### Access Modifier in Python
- We can choose to provide access to our variable by making them public or private.
- Normally , we make all the data members private.

```python
class Student:
    numberOfStudents = 0  # Class variable for tracking the number of students
    SchoolName = "Earth"  # Class variable for the school name

    def __init__(self, name, rollNumber, marks):
        self.name = name
        self.rollNumber = rollNumber
        self.__marks = marks
        Student.numberOfStudents += 1  # Increment the class variable when a new student is created
    
    ## Getter
    def getMarks(self):
        return self.__marks
    ## Setter
    def setMarks(self,newMarks):
        self.__marks = newMarks
```

```python
s1 = Student('Goku',1,50)
```

```python
s1.setMarks(90)
```

```python
s1.getMarks()
```
