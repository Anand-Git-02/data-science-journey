## 1. What is PyTorch?

**PyTorch** is an open-source machine learning and deep learning framework primarily developed by Meta's AI Research lab. Based on the Torch library, it is highly flexible, easy to debug, and feels very "Pythonic." It allows researchers and developers to build dynamic neural networks that can be modified on the fly.

## Who Uses PyTorch?

PyTorch is an industry standard used by top technology companies to power major AI products and services:

* **Meta Platforms (Facebook):** Developed PyTorch; uses it extensively for computer vision, NLP, and AI research across Facebook and Instagram.
* **OpenAI:** Utilizes PyTorch for training large-scale language models and generative models, including GPT Models, ChatGPT, and DALL·E.

## 2. Core Features

PyTorch's architecture is built around several key features that make it a favorite among data scientists:

1. **Tensor Computations:** Multi-dimensional arrays similar to NumPy, but heavily optimized for deep learning.
2. **GPU Acceleration:** Tensors and mathematical operations can be easily moved to GPUs for massive speedups.
3. **Dynamic Computation Graph:** PyTorch builds the mathematical graph as your code runs (define-by-run), making debugging incredibly intuitive.
4. **Automatic Differentiation:** An engine that automatically computes derivatives (gradients) for all tensor operations, which is essential for training models.
5. **Distributed Training:** Natively supports scaling training jobs across multiple GPUs and machines.
6. **Interoperability with other libraries:** Blends seamlessly with Python's data science stack (like NumPy and SciPy).

## 5. Understanding Tensors in PyTorch

### What are Tensors?
At its core, a Tensor is a specialized multi-dimensional array designed specifically for mathematical and computational efficiency. You can think of them as the primary data structure in deep learning. While they are very similar to standard arrays, tensors are optimized to track dimensions (directions/span) and run on specialized hardware.

#### Real-World Examples of Tensors by Dimension:

<img alt="tensor_dimensions" caption="Pytorch tensor_dimensions" src="../Image/tensor_dimensions.png"> 

1. **Scalars (0-dimensional tensors):** A single number.
* *Usage:* Represents a single value, often used for simple metrics or constants.
* *Example:* **Loss value**. After a forward pass, the loss function computes a single scalar value indicating the difference between the predicted and actual outputs (e.g., `5.0` or `-3.14`).


2. **Vectors (1-dimensional tensors):** A list of numbers (arrays).
* *Usage:* Represents a sequence or a collection of values.
* *Example:* **Word Embeddings / Feature Vectors**. In Natural Language Processing (NLP), each word in a sentence may be represented as a 1D vector. (e.g., `[0.12, -0.84, 0.33]` from a pre-trained model like Word2Vec or GloVe).


3. **Matrices (2-dimensional tensors):** A 2D grid of numbers.
* *Usage:* Represents tabular or grid-like data.
* *Example:* **Grayscale Images**. A grayscale image can be represented as a 2D tensor, where each entry corresponds to the pixel intensity. (e.g., `[[0, 255, 128], [34, 90, 180]]`).


4. **3D Tensors:** Adds a third dimension, often used for stacking data.
* *Usage:* **Coloured Images (RGB)**.
* *Example:* A single RGB image is represented as a 3D tensor mapping `(width × height × channels)`. An image of size 256x256 with 3 color channels (Red, Green, Blue) has the shape `[256, 256, 3]`.


5. **4D Tensors:** Adds the batch size as an additional dimension to 3D data.
* *Usage:* **Batches of RGB images**.
* *Example:* A dataset of coloured images is represented as a 4D tensor mapping `(batch size × width × height × channels)`. A batch of 32 images, each size 128x128 with 3 color channels, has the shape `[32, 128, 128, 3]`.


6. **5D Tensors:** Adds a time dimension for data that changes over time.
* *Usage:* **Video data**. Video clips are represented as a sequence of frames, where each frame is an RGB image.
* *Example:* A batch of 10 video clips, where each clip has 16 frames of size 64x64 and 3 color channels, would have the shape `[10, 16, 64, 64, 3]`.

## 6. Why Are Tensors Useful?

1. **Mathematical Operations:** Tensors enable highly efficient mathematical computations (addition, multiplication, dot products, etc.) that are strictly necessary for neural network operations.
2. **Representation of Real-world Data:** Almost any data type can be numerically represented as a tensor:
* **Images:** 3D tensors `(width × height × channels)`.
* **Text:** Tokenized and represented as 2D or 3D tensors `(sequence length × embedding size)`.
* **Audio & Video:** Represented with time and frequency dimensions.


3. **Efficient Computations:** Tensors are explicitly optimized for hardware acceleration. This allows massive parallel computations on **GPUs** (Graphics Processing Units) or **TPUs**, which is absolutely crucial for training deep learning models in a reasonable amount of time.

- `Reshape()` changes the dimensions (shape) of an array or tensor without altering its underlying data or the total number of elements.

```python
input_shape =  torch.randint(0,100,(9,))
print(f"{input_shape} \n {input_shape.shape}")

input_shape.reshape((3,3))
```

- `squeeze()` Removes any dimension that has a size of 1, which commonly comes up after batch processing when you have a leftover single-item dimension you no longer need.

```python
# 1. Create a "batch" containing one result with 10 classes
# Shape: (1, 10) -> [Batch, Classes]
output = torch.randn(1, 10)
print(f"Original Shape: {output.shape}")


# 2. Squeeze it to remove the batch dimension
# This turns it into a simple 1D vector of 10 elements
squeezed_output = output.squeeze()
print(f"Squeezed Shape: {squeezed_output.shape}")
```

- `unsqueeze()` inserts a new dimension of size 1 at a position you specify, which is often needed to match the shape a model or function expects, such as adding a batch dimension to a single image.

```python
# 1. Simulate a single RGB image (3 channels, 224x224 pixels)
# Shape: [3, 224, 224]
image = torch.randn(3, 224, 224)
print(f"Original Image Shape: {image.shape}")

# 2. Add the batch dimension at index 0
# This makes it a "batch of 1"
# Shape: [1, 3, 224, 224]
input_tensor = image.unsqueeze(0)
print(f"Model-Ready Shape:    {input_tensor.shape}")
```

- `flatten()` collapses all the dimensions of a tensor into a single long 1D vector, which is the standard step performed just before passing data into a fully connected (linear) layer.

```python
input_dimension = torch.randint(1,10,(2,2))
print(input_dimension)

input_dimension.flatten()
```

- `permute()` rearranges the original dimensions of a tensor into a specific new order, allowing you to "shuffle" the axes themselves rather than just changing the numbers in the shape.

    - **The Scenario:** Changing Image Formats
      This is most common when moving between image libraries. Many visualization tools (like Matplotlib) expect images to be Channel-Last (Height, Width, Channels), while PyTorch models expect Channel-First (Channels, Height, Width)

```python
# 1. Start with a Channel-Last image (e.g., from OpenCV or PIL)
# Shape: [224, 224, 3] -> (H, W, C)
image = torch.randn(224, 224, 3)
print(f"Original Shape (H, W, C): {image.shape}")

# 2. Permute to Channel-First for a PyTorch model
# We want: 
# Original Dim 2 (Channels) to be at index 0
# Original Dim 0 (Height) to be at index 1
# Original Dim 1 (Width) to be at index 2
permuted_image = image.permute(2,1,0)

print(f"Permuted Shape (C, H, W): {permuted_image.shape}")
```

## 12. Moving Tensors to a Device (CPU / GPU)

#### Moving Tensors to a Device (CPU / GPU)

In PyTorch, a **device** refers to the physical hardware where a tensor's data lives and where its computations are executed. This can be either:
- **CPU** (Central Processing Unit) - Default, works everywhere
- **GPU** (Graphics Processing Unit) - CUDA-enabled NVIDIA GPUs for massive speedups
- **MPS** (Metal Performance Shaders) - Apple Silicon (M1/M2/M3) acceleration

**Why this matters:**
- Neural network training on a GPU can be 10-100x faster than on CPU
- A tensor can only interact with other tensors on the **same device**
- All PyTorch operations require their input tensors to be on the same device

**Critical rule:** Tensors created on one device stay on that device until you explicitly move them.

```python
import torch

# Check PyTorch version
print(f"PyTorch version: {torch.__version__}")
print()

# Check if CUDA (NVIDIA GPU) is available
if torch.cuda.is_available():
    print("✓ CUDA is available!")
    print(f"  GPU Device: {torch.cuda.get_device_name(0)}")
    print(f"  Number of GPUs: {torch.cuda.device_count()}")
else:
    print("✗ CUDA not available - using CPU")

print()

# Check if MPS (Apple Silicon) is available
if torch.backends.mps.is_available():
    print("✓ MPS (Apple Silicon) is available!")
else:
    print("✗ MPS not available")
```
