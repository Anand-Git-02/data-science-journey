```python
import torch
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np

# Set up the figure and grid layout
fig = plt.figure(figsize=(14, 10))
fig.suptitle("Visualizing Tensor Dimensions (Ranks)", fontsize=20, fontweight='bold', y=0.95)

# Define visual style matching your notes
box_color = '#d0e3f4'  # Light blue
edge_color = '#5b9bd5' # Darker blue border
line_width = 1.5

# ==========================================
# 1. Rank 0: Scalar (0-dimensional)
# ==========================================
scalar_tensor = torch.tensor(7) # A single scalar value

ax0 = fig.add_subplot(221)
# Draw a single square
ax0.add_patch(Rectangle((0.5, 0.5), 1, 1, facecolor=box_color, edgecolor=edge_color, lw=line_width))
ax0.set_xlim(0, 2); ax0.set_ylim(0, 2)
ax0.axis('off')
ax0.set_title(f"Rank 0: Scalar\nPyTorch Shape: {list(scalar_tensor.shape)}", fontsize=14)


# ==========================================
# 2. Rank 1: Vector (1-dimensional)
# ==========================================
# Create a 1D tensor of length 5
vector_tensor = torch.ones(5) 

ax1 = fig.add_subplot(222)
# Draw 5 squares in a row
for i in range(vector_tensor.shape[0]):
    ax1.add_patch(Rectangle((i * 1.1, 0.5), 1, 1, facecolor=box_color, edgecolor=edge_color, lw=line_width))
ax1.set_xlim(-0.5, 6); ax1.set_ylim(0, 2)
ax1.axis('off')
ax1.set_title(f"Rank 1: Vector\nPyTorch Shape: {list(vector_tensor.shape)}", fontsize=14)


# ==========================================
# 3. Rank 2: Matrix (2-dimensional)
# ==========================================
# Create a 2D tensor (3 rows, 4 columns)
matrix_tensor = torch.ones(3, 4) 

ax2 = fig.add_subplot(223)
# Draw a grid of 3x4 squares
for i in range(matrix_tensor.shape[1]):     # Columns
    for j in range(matrix_tensor.shape[0]): # Rows
        # (2-j) inverses the y-axis so it builds top-to-bottom
        ax2.add_patch(Rectangle((i * 1.1, (2 - j) * 1.1), 1, 1, facecolor=box_color, edgecolor=edge_color, lw=line_width))
ax2.set_xlim(-0.5, 5); ax2.set_ylim(-0.5, 4)
ax2.axis('off')
ax2.set_title(f"Rank 2: Matrix\nPyTorch Shape: {list(matrix_tensor.shape)}", fontsize=14)


# ==========================================
# 4. Rank 3: 3D Tensor (3-dimensional)
# ==========================================
# Create a 3D tensor (3 deep, 3 rows, 4 columns)
tensor_3d = torch.ones(3, 4, 3) 

ax3 = fig.add_subplot(224, projection='3d')
# Matplotlib's voxel plot requires a boolean array to know where to draw cubes
voxel_grid = np.ones(tensor_3d.shape, dtype=bool)

# Plot the 3D cubes
ax3.voxels(voxel_grid, facecolors=box_color, edgecolors=edge_color, alpha=0.9, linewidth=1)
ax3.axis('off')
ax3.set_title(f"Rank 3: 3D Tensor\nPyTorch Shape: {list(tensor_3d.shape)}", fontsize=14)

# Adjust layout and display
plt.tight_layout(rect=[0, 0.03, 1, 0.90])
plt.show()
```
