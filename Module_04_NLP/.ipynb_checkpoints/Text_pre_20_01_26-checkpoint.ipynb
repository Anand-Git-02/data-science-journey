#### 2. Regular Expression

```python
import re
sent3 = 'I am going to delhi!'

tokens = re.findall("[\w']+",sent3)
tokens
```

```python
# ->datastes --> lower() --> html tag remove --> punctuation --> chatwords --> emojies --> sentance words (nltk) --> steamming


sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations="?:!.,;"

import nltk
from nltk.stem import WordNetLemmatizer

wordnetlemmma = WordNetLemmatizer()


Sentace_words = nltk.word_tokenize(sentence)


for word in Sentace_words:
    if word in punctuations:
        Sentace_words.remove(word)
# print(Sentace_words)

print("{0:30}{1:30}".format("Words","Lemma"))

for i in Sentace_words:
    print("- {0:30} - {1:30}".format(i,wordnetlemmma.lemmatize(i,pos='v')))
```
