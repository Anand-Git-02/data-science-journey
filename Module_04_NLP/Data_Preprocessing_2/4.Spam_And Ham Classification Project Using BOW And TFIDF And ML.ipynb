### Step 1 :- Load Dataset

```python
import pandas as pd
df=pd.read_csv('Data/SMSSpamCollection.txt',
                    sep='\t',names=["label","message"])
```

```python
# pd.set_option('display.max_rows',None)
# pd.set_option('display.max_colwidth', None)
```

```python
df.head()
```

### Step : 2 Data Cleaning
- Stop Remove
- Stemming

```python
## Data Cleaning And Preprocessing
import re
import nltk
# nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# stopwords.words('english')
```

```python
## Q&A,chatbots,text summarization
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import re
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-z]',' ',df['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
# corpus
```

```python
df.head()
```

### Step 3 Data Preprocessing
- Change Categorical Output Column into Numberical Column

```python
## Output Features
y=pd.get_dummies(df['label'],dtype=int)
y=y.iloc[:,0].values
y
```

-----
Train Model From Here.

```python
## Train Test Split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
len(X_train) , len(X_test)
```

```python
len(y_train) , len(y_test)
```

### Create the Bag OF Words model

```python
from sklearn.feature_extraction.text import CountVectorizer

## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
## independent features
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

### Step : 2 Creating The TF-IDF Model for Text to Vectors

### Step : 3 Creating Machine Learning Model for Train Datasets

```python
#prediction
y_pred=spam_tfidf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```
