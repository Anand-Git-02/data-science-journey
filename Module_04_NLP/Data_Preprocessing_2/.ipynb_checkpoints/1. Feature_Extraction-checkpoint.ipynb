```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Input dictionary
dic = {
    'text': ['he is a good boy', 'she is a good girl', 'boy and girl are good'],
    'output': [1, 1, 1]
}
```

```python
# Convert dictionary to DataFrame
df = pd.DataFrame(dic)
```

```python
df
```

```python
# Define stop words
stop_words = set(stopwords.words('english'))
```

```python
# Function to remove stop words
def remove_stop_words(text):
    words = word_tokenize(text)  # Tokenize the text
    filtered_words = [word for word in words if word.lower() not in stop_words]  # Filter out stop words
    return ' '.join(filtered_words)  # Join words back into a string
```

```python
# Apply the function to the 'text' column
df['text'] = df['text'].apply(remove_stop_words)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=3)
```

```python
bow = cv.fit_transform(df['text'])
```

```python
# vocab

print(cv.vocabulary_)
```

```python
sorted(cv.vocabulary_)
```

```python
cv.transform(['good girl']).toarray()
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
cv.transform(['good boy jack']).toarray()
```
