##### NER detects `entities` and labels them into predefined categories such as:
1. Person
2. Organization
3. Location
4. Date / Time
5. Money
6. Medical / Scientific entities (domain-specific)

- Converts Unstructured Text into Structured Data
- Improves Information Extraction
- Critical for Search & Recommendation Systems
- Enhances Question Answering Systems

Named Entity Recognition (NER) identifies real-world entities like people, organizations, locations, dates, and amounts in text.
It converts unstructured language into structured, machine-understandable information for tasks like search, QA, and analytics.

```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
tag_elements=nltk.pos_tag(words)
tag_elements
```

```python
nltk.download('maxent_ne_chunker')
```

```python
nltk.download('words')
```

```python
nltk.ne_chunk(tag_elements).draw()
```
