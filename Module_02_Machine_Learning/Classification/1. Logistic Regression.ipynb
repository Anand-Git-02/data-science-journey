### 1. Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_wine
```

```python
import warnings
warnings.filterwarnings('ignore')
```

#### Reading Data

```python
data = load_wine()
```

### 2. Understanding the Dataset

```python
print('Descripation of Dataset:',data.DESCR)
```

```python
print("Keys in the dataset:")
for feature in data.keys():
    print(f" - {feature}")
```

#### Independent Variable and Dependent Variable

```python
X = data['data']
y = data['target']
print("Dimension of Independent Variable:",X.shape)
print("Dimension of Dependent   Variable:",y.shape)
```

#### Features Name

```python
print("Feature Names:")
for feature in data.feature_names:
    print(f" - {feature}")
```

#### i) Convert it into DataFrame

```python
data['target_names']
```

```python
df1 = pd.DataFrame(data['data'],columns=data['feature_names'])
df2 = pd.DataFrame(data['target'],columns=['Class'])
```

```python
df = pd.concat([df1,df2],axis=1)
df.head()
```

#### ii) Determining missing values

```python
print('Missing Values:\n')
print(df.isnull().sum())
```

### 3. Creating training , test dataset for  Dependent and Independent vaiable

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=30)
```

```python
print('Dimension of Training Dataset :-',X_train.shape)
print('Dimension of Test     Dataset :-',X_test.shape)
```

### 4. Model Building

```python
from sklearn.linear_model import LogisticRegression
```

```python
Model_Lr = LogisticRegression()
```

##### Fitting the Model

```python
Model_Lr.fit(X_train,y_train)
```

##### Determining the Scores of Training and Testing Datasets

```python
print(f"Training set score :- {Model_Lr.score(X_train,y_train)}")
print(f"Testing  set score :- {Model_Lr.score(X_test,y_test)}")
```

### 5. Predicting and determining accuracy of Model.

```python
y_pred = Model_Lr.predict(X_test)
y_pred
```

### 6. Performance Metrics  

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
print(f'Results of the confusion Matrix :-\n {cm}')
```

```python
print(f'Classification Report :-\n{classification_report(y_test,y_pred)}')
```

```python
print(f'Accuracy :- {accuracy_score(y_test,y_pred)}')
```
