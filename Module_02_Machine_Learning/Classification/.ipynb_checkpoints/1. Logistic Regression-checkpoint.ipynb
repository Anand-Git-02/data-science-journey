```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
```

```python
df = pd.read_csv(r"Social_Network_Ads.csv")
df
```

```python
df.isnull().sum()
```

```python
df['Purchased'].value_counts()
```

```python
X = df.drop(columns='Purchased')
y = df['Purchased']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import RobustScaler
rc = RobustScaler()
X[['Age','EstimatedSalary']] = rc.fit_transform(X[['Age','EstimatedSalary']])
```

```python
X
```

# train test split

```python
X.shape
```

```python
(400 * 20)/100
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state =50)
```

```python
X_train.shape , X_test.shape
```

```python
X_test
```

```python
y_test
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

# Prediction

```python
X_test
```

```python
df.iloc[[332]]
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
y_test.values
```

# evalution

```python
y_test.value_counts()
```

```python
from sklearn.metrics import classification_report,confusion_matrix , accuracy_score 
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn import metrics
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels =['1','0'])
cm_display.plot()
```

```python
accuracy_score(y_test,y_pred)
```

```python
y_test.shape
```

```python
print(classification_report(y_pred,y_test))
```

```python
sns.heatmap(cm,annot=True)
```

## Hyperparameter Tuning And Cross Validation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Define model
model_h = LogisticRegression(max_iter=1000)

# Define hyperparameters
tuned_parameters = {
    'penalty': ['l1', 'l2', 'elasticnet', None],
    'C': [100, 10, 1.0, 0.1, 0.01],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
}
```

```python
# Perform Grid Search
grid_search = GridSearchCV(model_h, tuned_parameters, cv=5, n_jobs=-1, scoring='accuracy')

# Fit the model (Assuming X and y are your features and labels)
grid_search.fit(X, y)
```

```python
# Display the best parameters and best score
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Score: {grid_search.best_score_}")
```
