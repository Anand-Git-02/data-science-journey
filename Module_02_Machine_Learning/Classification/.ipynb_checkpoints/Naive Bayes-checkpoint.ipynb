### 1. Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

#### Reading Data

```python
Training = pd.read_csv('Data/datatraining.txt')
Test = pd.read_csv('Data/datatest.txt')
```

```python
Training.head()
```

### 2. Data Overview

##### i) Number of Rows and Columns

```python
print(f"Train Data → Rows: {Training.shape[0]}, Columns: {Training.shape[1]}")
print(f"Test Data  → Rows: {Test.shape[0]}, Columns: {Test.shape[1]}")
```

##### ii) The Name of Variables in Dataset

```python
Training.columns
```

```python
Training.isna().sum()
```

```python
Test.isna().sum()
```

```python
Training.drop('date',axis=1,inplace=True)
Test.drop('date',axis=1,inplace=True)
```

### 3. Creating Training and Test for Independent and Depeadent Variables

```python
# Train Data
X_train = Training.drop('Occupancy',axis=1)
y_train = Training['Occupancy']

# Test Data
X_test = Test.drop('Occupancy',axis=1)
y_test = Test['Occupancy']
```

### 4. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### 5. Model Buliding

```python
from sklearn.naive_bayes import GaussianNB

Model_nav = GaussianNB()
Model_nav.fit(X_train,y_train)
```

```python
# Model Score

print('Accuracy of Naive Bayes Model on Training Set :- %0.4f'%Model_nav.score(X_train,y_train))
print('Accuracy of Naive Bayes Model on Test     Set :- %0.4f'%Model_nav.score(X_test,y_test))
```

### 6. Model Evaluation

```python
from sklearn.metrics import accuracy_score ,confusion_matrix , classification_report
```

```python
y_prad = Model_nav.predict(X_test)
y_prad
```

##### i) Confusion Matrix

```python
confusion_matrix(y_test,y_prad)
```

##### ii) Classification Report

```python
print(classification_report(y_test,y_prad))
```
