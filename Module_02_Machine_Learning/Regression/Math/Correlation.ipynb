```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
dict_1 = {
    'Students' : [i for i in range(1,6)],
    'Math(x)'     : [5,7,9,3,6],
    "Science(y)"   : [6,6,8,1,5]
}
```

```python
# Creating Dataframe 

df = pd.DataFrame(dict_1)
df
```

```python
df.columns
```

```python
df_corr = df[['Math(x)', 'Science(y)']]
```

```python
# Correlation
df_corr.corr()
```

```python
df_corr
```

# Correlation
#### **Step 1 :-** Find the Sum of (`Math(x)`,`Science(y)`)



```python
X_sum = df_corr['Math(x)'].sum()
X_sum
```

```python
Y_sum = df_corr['Science(y)'].sum()
Y_sum
```

#### **Step 2 :-** `Mutiple X and Y and Find the XY Sum`

```python
XY = df_corr['Math(x)'] * df['Science(y)']
XY
```

```python
XY_sum = XY.sum()
XY_sum
```

#### **Step 3 :-** `Find the X square and Y Square and Sum separately`

```python
X_square = df['Math(x)'] ** 2
X_square
```

```python
X_square_sum = X_square.sum()
X_square_sum
```

```python
Y_Square = df['Science(y)'] ** 2
Y_Square
```

```python
Y_square_sum = Y_Square.sum()
Y_square_sum
```

#### **Step 4 :-** Find the Mean

```python
df_corr.mean()
```

```python
X_mean = X_sum/df_corr.shape[0]
X_mean
```

```python
Y_mean = Y_sum/df_corr.shape[0]
Y_mean
```

#### **Step 5:-** Finding



```python
Covariance = (XY_sum/df_corr.shape[0]) - (X_mean * Y_mean)
Covariance
```

#### Step 6:- Finding



```python
df_corr.std()
```

```python
Variace_X = (X_square_sum/df_corr.shape[0]) - (X_mean ** 2)
Std_X = np.sqrt(Variace_X)
Std_X
```

```python
Variace_Y = (Y_square_sum/df_corr.shape[0]) - (Y_mean ** 2)
Std_Y = np.sqrt(Variace_Y)
Std_Y
```

#### Step 7 :-


```python
Correlation = Covariance/(Std_X*Std_Y)
Correlation
```
